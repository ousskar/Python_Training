## What is Python?

Python is a high-level programming language created by **Guido van Rossum**. It emphasizes readability and simplicity: you can usually express an idea in fewer lines of code than in languages such as C, C++, or Java.

As of the **January 2026 TIOBE Index**, Python ranks **#1** worldwide with a rating of **22.61%**, ahead of C, Java, and C++.

Python is popular because it is useful across many domains:

- Data science and analytics
- Artificial intelligence and machine learning
- Web development and APIs
- Automation and scripting
- Scientific computing and education

That combination of readability, ecosystem, and versatility is why Python is a strong first language — and why it remains a core tool for professional work.

## A short history of Python

You do not need to memorize these dates, but the milestones help explain how Python became what it is today:

1. **December 1989** — Guido van Rossum started designing a new scripting language during the Christmas holiday. He intended it as a successor to the ABC language, and named it after *Monty Python's Flying Circus*.
2. **February 1991** — The first public release (version 0.9.0) was posted to the `alt.sources` newsgroup.
3. **January 1994** — Python 1.0 was released.
4. **October 2000** — Python 2.0 introduced a more open development process and helped the ecosystem grow.
5. **December 2008** — Python 3.0 was released. It modernized the language but was not fully backward compatible with Python 2.
6. **April 2011** — `pip` was released, giving Python a standard package manager.
7. **July 2018** — Guido stepped down as "Benevolent Dictator for Life" (BDFL).
8. **January 2020** — Official support for Python 2 ended. New work should use **Python 3**.
9. **Today** — Python powers large parts of AI, computer vision, recommendation systems, data science, automation, testing, and more — supported by a large open-source ecosystem.

> **Course note:** This training uses **Python 3.10+**. Do not install or use Python 2.

## Advantages and trade-offs

### Advantages

1. **Simple and readable** — Easier to learn than many other languages.
2. **High productivity** — You can often do more with less code.
3. **Open source, strong community** — Huge library ecosystem and lots of learning resources.
4. **Versatile** — Useful for scripting, apps, data work, AI, and automation.
5. **Great "glue" language** — Works well with code and tools written in other languages.
6. **Cross-platform** — The same script usually runs on Windows, macOS, and Linux.

### Main trade-off

Python is typically **slower than compiled languages** such as C, C++, or Go. For most day-to-day applications, data scripts, and backend services, that trade-off is acceptable. When raw execution speed is critical (for example, low-level systems code), those other languages may be a better fit.

## Installing the Python interpreter

To run Python programs, you need an **interpreter** on your computer. We recommend **CPython**, the official Python 3 implementation.

1. Visit the [Python downloads page](https://www.python.org/downloads/).
2. Click **Download** for the latest Python 3 release.
3. Choose the installer for your operating system. On Windows, pick the **64-bit** installer marked **Recommended**, as shown below:

![Python download options](images/3.png)

### Windows installation walkthrough

The steps below use **Windows** as the main example (matching the screenshots in this notebook). macOS and Linux users can install from [python.org](https://www.python.org/downloads/) or through their package manager; the verification commands are the same.

Double-click the installer you downloaded. You should see a setup screen similar to this:

![Python installer start screen](images/4.png)

#### 1. Add Python to PATH

On the first screen, check **Add python.exe to PATH**.

- **Why it matters:** This lets you run `python` and `pip` from any terminal, without typing the full install path.

#### 2. Use admin privileges for `py.exe`

Also select **Use admin privileges when installing py.exe**.

- **Why it matters:** It reduces permission problems later when installing tools or packages.

#### 3. Prefer Customize Installation

Choose **Customize Installation** instead of the default "Install Now".

- Default installs often hide Python deep under `AppData`.
- Custom install lets you review optional features and choose a clear install location (for example `C:\Python313`).

![Optional features](images/5.png)

![Advanced options / install location](images/6.png)

![Installation complete](images/7.png)

### Verifying your installation

After setup finishes, open a terminal:

- **Windows:** Command Prompt or PowerShell
- **macOS / Linux:** Terminal

Then run:

```bash
python --version
```

You should see something like `Python 3.13.x` (any 3.10+ version is fine for this course).

Also check that `pip`, Python's package manager, is available:

```bash
pip --version
```

![Successful python and pip version checks](images/8.png)

> **If `python` is not recognized:** Close and reopen the terminal, then try again. If it still fails, reinstall and make sure **Add python.exe to PATH** was checked.
>
> **If some scientific packages fail to install later:** You may need Microsoft's [Visual Studio Build Tools](https://visualstudio.microsoft.com/visual-cpp-build-tools/). This is *not* required for installing Python itself — only for certain packages that compile native code.

## Writing and running Python code

In this course you will mostly work in **Jupyter notebooks**, but it helps to know the three common ways to run Python:

| Method | When to use it |
| --- | --- |
| **Interactive shell (REPL)** | Quick experiments in the terminal |
| **Script file (`.py`)** | Small programs and reusable tools |
| **Jupyter notebook (`.ipynb`)** | Learning, exploration, and this course |

### Try it here

Run the cell below. It should print a greeting.

````python
print("Hello, Python!")
print("2 + 3 =", 2 + 3)
````

### Run the same idea as a script

1. Create a file named `hello.py` with this content:

```python
print("Hello, Python!")
print("2 + 3 =", 2 + 3)
```

2. In a terminal, move to the folder that contains the file, then run:

```bash
python hello.py
```

You should see the same output as in the notebook cell above.

### Optional: try the interactive shell

In a terminal, type `python` and press Enter. You will see the `>>>` prompt. Try:

```python
>>> print("Hello from the REPL")
>>> 10 * 4
```

Type `exit()` or press `Ctrl+Z` then Enter (Windows) / `Ctrl+D` (macOS/Linux) to leave the shell.
