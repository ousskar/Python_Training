# 1. Modules and Packages

A module is a Python file that groups related functions, classes, and variables. In this notebook you will learn how to import modules, inspect what they contain, understand how Python searches for them, and see how packages organize larger projects.

```python
from pathlib import Path
import sys

module_dir = Path.cwd()
if not (module_dir / "fibo.py").exists():
    module_dir = Path("08 - Python Libraries, Modules, and Packages")

if str(module_dir.resolve()) not in sys.path:
    sys.path.insert(0, str(module_dir.resolve()))

print("Module directory:", module_dir.resolve())
```

```python
import fibo
# Access the function as follows
fibo.fib(50)
```

```python
fibo.fib2(8)
```

You can also import selected names from a module directly into the current namespace. This makes calls shorter, but it also hides where those names came from.

```python
from fibo import fib, fib2
fib(50)
```

If you call a function frequently, you can bind it to a shorter local name. Use this sparingly so the code stays readable.

```python
fib = fibo.fib
fib(100)
```

You can also import all public names from a module with `from fibo import *`, but this is usually discouraged because it can overwrite existing names and makes code harder to read.

```python
from fibo import *

fib(20)
fib2(20)
```

You can rename a module during import with `as`. This is useful when the original name is long or when a shorter alias is a common convention.

```python
import fibo as fib

fib.fib(50)
```

```python
from fibo import fib as fibonacci
fibonacci(50)
```

## 2. Executing a Module as a Script

From a terminal, you can run the module file directly:

`python fibo.py 50`

In this example, `50` becomes the command-line argument that the script reads from `sys.argv`.

When a module is run directly, Python sets `__name__` to `"__main__"`. This lets you keep reusable functions at the top of the file and place script-only behavior behind a guard at the bottom.

```python
if __name__ == "__main__":
    import sys
    fib(int(sys.argv[1]))
```

# 5. Packages

Packages organize related modules into directories so larger programs stay structured.

A dotted name such as `sound.effects.echo` means:

- `sound` is the top-level package
- `effects` is a subpackage inside `sound`
- `echo` is a module inside `sound.effects`

![alt text](hier.png "Title")

When importing a package, Python searches the directories listed in `sys.path`.

A file named `__init__.py` marks a directory as a package in the classic package model and can also run package initialization code.

The examples below are illustrative. The `sound` package does not exist in this training workspace, so they explain syntax rather than provide runnable code here.

```python
# Example only: the sound package is illustrative and is not part of this workspace.
import sound.effects.echo
```

This import loads the module, but you still access it through its full package path:

`sound.effects.echo.echofilter(input, output, delay=0.7, atten=4)`

If you want to use `echo` without the full prefix, import it directly with `from sound.effects import echo`.

Another option is to import the function you need directly from the module:

```python
# Example only: direct import from a module inside a package.
from sound.effects.echo import echofilter
echofilter(input, output, delay=0.7, atten=4)
```

## 5.1 Importing `*` from a Package

If a package's `__init__.py` file defines a list named `__all__`, Python uses that list to decide which names to import when `from package import *` is used. Package authors must keep `__all__` up to date as the package evolves.

```python
__all__ = [
    "echo",      # refers to the 'echo.py' file
    "surround",  # refers to the 'surround.py' file
    "reverse",   # !!! refers to the 'reverse' function now !!!
]
```

## 7.1 Import Best Practices

Use imports in a consistent way so your code stays readable and easier to debug.

Recommended habits:

- keep imports near the top of the file
- group imports logically: standard library, third-party packages, then local modules
- prefer explicit imports such as `from fibo import fib2` when you only need a small number of names
- avoid `from module import *` in production code because it hides where names came from
- use clear aliases only when they improve readability, such as `import numpy as np`
- prefer absolute imports in larger projects because they are easier to follow
- use relative imports carefully inside packages when they make the package structure clearer
- avoid running heavy code during import time; define functions and classes first, then call them later
- keep script-only code inside `if __name__ == "__main__":` so importing the module does not trigger unwanted behavior

A useful pattern is: imports first, reusable definitions next, and execution logic last.
