# 3. OOP and Decorators Exercises Solutions

This notebook provides sample implementations for the OOP, dataclass, higher-order function, and decorator exercises.

```python
class Car:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed

    def describe(self):
        return f'{self.brand} is moving at {self.speed} km/h'

    @property
    def is_fast(self):
        return self.speed > 100

    def __str__(self):
        return self.describe()

class BankAccount:
    def __init__(self, balance=0):
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if amount > self.balance:
            raise ValueError('Insufficient funds')
        self.balance -= amount
```

```python
from dataclasses import dataclass

class Person:
    def __init__(self, name):
        self.name = name

    def describe(self):
        return f'Person: {self.name}'

class Employee(Person):
    def __init__(self, name, role):
        super().__init__(name)
        self.role = role

    def describe(self):
        return f'{super().describe()}, role={self.role}'

@dataclass(frozen=True)
class Product:
    name: str
    price: float = 0.0

    def __post_init__(self):
        if self.price < 0:
            raise ValueError('price must be non-negative')
```

```python
from functools import wraps
import time

def apply_to_each(values, func):
    return [func(value) for value in values]

def make_multiplier(multiplier):
    def inner(value):
        return value * multiplier
    return inner

def announce(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        print(f'Running {func.__name__}')
        return func(*args, **kwargs)
    return wrapper

def timed(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        print(time.time() - start)
        return result
    return wrapper

@announce
@timed
def greet(name):
    return f'Hello, {name}'
```
