## Exercise 1 – Basic Function Types

**Original task**

Add type hints to:

```python
def repeat(text, times):
    return text * times
```

Requirements:
- `text: str`
- `times: int`
- Return type: `str`

Then call `repeat("hi", 3)` and `repeat(123, 3)` and check what a type checker reports.


````python
def repeat(text: str, times: int) -> str:
    """Repeat a string a given number of times.

    A type checker will warn if text is not str or times is not int.
    """
    return text * times

print(repeat("hi", 3))       # OK
print(repeat("abc", 1))      # OK

# A static type checker (mypy/pyright) should warn here:
# print(repeat(123, 3))
````

## Exercise 2 – Collections

**Original task**

Annotate:

```python
def average(values):
    return sum(values) / len(values)
```

Requirements:
- `values` should be something like `list[int]` or `list[int | float]`.
- Return type: `float`.

Test with numeric lists, then see what a type checker says for invalid types.


````python
from typing import Iterable

Number = int | float

def average(values: Iterable[Number]) -> float:
    """Compute the average of a non-empty iterable of numbers.

    For simplicity, we don't handle the empty case here.
    """
    total = 0.0
    count = 0
    for v in values:
        total += float(v)
        count += 1
    return total / count

print(average([1, 2, 3]))            # 2.0
print(average([1.0, 2.5, 3.0]))      # 2.166...

# A static type checker will flag this as an error:
# print(average(["a", "b"]))
````

## Exercise 3 – Optional and Union (`parse_int`)

**Original task**

Implement and annotate:

```python
def parse_int(value):
    ...
```

Requirements:
- Input: `value: str | int`
- Output: `int | None`

Behavior:
- If `value` is `int`, return it.
- If `value` is `str` and convertible with `int(value)`, return the result.
- Otherwise, return `None`.


````python
from typing import Union

def parse_int(value: int | str) -> int | None:
    """Parse an int from either an int or a string.

    Returns None if conversion is not possible.
    """
    if isinstance(value, int):
        return value
    # Now value is str
    try:
        return int(value)
    except ValueError:
        return None

print(parse_int(10))       # 10
print(parse_int("20"))     # 20
print(parse_int("abc"))    # None
````

## Exercise 4 – Generic Box

**Original task**

Implement:

```python
T = TypeVar("T")

class Box(Generic[T]):
    ...
```

Requirements:
- `__init__(value: T) -> None`
- `get() -> T`
- `set(value: T) -> None`

Then use `Box[int]` and `Box[str]`.


````python
from typing import Generic, TypeVar

    
T = TypeVar("T")

class Box(Generic[T]):
    def __init__(self, value: T) -> None:
        self._value = value

    def get(self) -> T:
        return self._value

    def set(self, value: T) -> None:
        self._value = value

box_int = Box[int](123)
box_str = Box[str]("abc")

print(box_int.get() + 1)        # 124
print(box_str.get().upper())    # "ABC"

# A type checker will catch mistakes like:
# box_int.set("not an int")
````
