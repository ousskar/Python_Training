## Sample Answers

Exercises 1 to 8

- Example variables: `name = 'Alice'`, `age = 25`, `learning_python = True`
- `int('42') * 2` gives `84`
- `max([4, 9, 1, 7, 3])` gives the largest value
- Tuple example: `cities = ('Paris', 'London', 'Tokyo')`; second city is `cities[1]`
- Dictionary example: `book = {'title': '1984', 'author': 'Orwell', 'year': 1949}`
- `set([1, 2, 2, 3, 4, 4, 5])` removes duplicates
- Use `type(...)` to inspect values
- Swap with `a, b = b, a`

```python
name = 'Alice'
age = 25
learning_python = True

print(int('42') * 2)
print(max([4, 9, 1, 7, 3]))

cities = ('Paris', 'London', 'Tokyo')
print(cities[1])

book = {'title': '1984', 'author': 'Orwell', 'year': 1949}
print(book['author'])

print(set([1, 2, 2, 3, 4, 4, 5]))
print(type(3.14), type(True), type('hello'), type([1, 2]))

a, b = 10, 20
a, b = b, a
print(a, b)
```

Exercises 9 to 16

- Use indexing like `text[0]` and `text[-1]`
- Use slicing like `text[:5]`, `text[::2]`, and `text[::-1]`
- `banana`.count('a') counts occurrences
- String helpers such as `.title()`, `.split()`, and `.join(...)` solve the remaining tasks

```python
text = 'Programming'
print(text[0], text[-1])
print(text[:5])
print('abcdefghij'[::2])
print('Python'[::-1])
print('banana'.count('a'))
print('hello world'.title())
print('Python is fun to learn'.split())
print(' - '.join(['read', 'write', 'code']))
```

Exercises 17 to 25

- Use loops, `enumerate()`, `zip()`, and `iter()` for iterable practice
- Use an accumulator variable for summing values
- List comprehensions can replace simple loop-based list building
- Finish with an `if/elif/else` example and a simulated input loop

```python
for value in [1, 2, 3, 4, 5, 6]:
    if value % 2 == 0:
        print(value)

for index, color in enumerate(['red', 'green', 'blue']):
    print(index, color)

for name, score in zip(['Alice', 'Bob'], [15, 18]):
    print(name, score)

iterator = iter([10, 20, 30])
print(next(iterator), next(iterator))

total = 0
for number in range(1, 11):
    total += number
print(total)

squares = []
for number in range(1, 9):
    squares.append(number * number)
print(squares)
print([number * number for number in range(1, 9)])

value = -3
if value > 0:
    print('positive')
elif value < 0:
    print('negative')
else:
    print('zero')

inputs = ['hello', 'again', 'quit']
for user_input in inputs:
    if user_input == 'quit':
        break
    print(user_input)
```
