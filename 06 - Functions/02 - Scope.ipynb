# Variable Scope in Python

Scope means **where a variable can be accessed** in your program.

- A **local variable** is created inside a function and is available only inside that function.
- A **global variable** is created outside functions and can be read from many places in the module.
- Prefer passing values into functions and returning results instead of changing global variables.

```python
length = 10
width = 5

def rectangle_properties(length, width):
    area = length * width
    perimeter = 2 * (length + width)
    diagonal = (length ** 2 + width ** 2) ** 0.5
    return area, perimeter, diagonal

area, perimeter, diagonal = rectangle_properties(length, width)

print(f"Area: {area}")
print(f"Perimeter: {perimeter}")
print(f"Diagonal: {diagonal:.2f}")
```

## Reading a global variable

A function can read a global variable if there is no local variable with the same name. This can be useful for constants, but too many global dependencies make functions harder to test.

```python
course_name = "Python Training"

def show_course_name():
    print(course_name)  # Read a global variable.

show_course_name()
print(course_name)
```

## Local variables and shadowing

If a variable is assigned inside a function, it is local to that function by default. A local variable can have the same name as a global variable; this is called shadowing. The global value is not changed.

```python
status = "outside"

def show_status():
    status = "inside"  # Local variable shadows the global variable.
    print(status)

show_status()
print(status)
```

## Changing a global variable with `global`

The `global` keyword tells Python that assignments inside the function should target the global variable. This works, but it couples the function to outside state, so use it carefully.

```python
count = 10

def increment_global_count():
    global count
    count += 1
    return count

print(increment_global_count())
print(count)
```

## Prefer returning a new value

Most functions are easier to understand when they receive values as inputs and return new values as outputs. This avoids hidden changes to global state.

```python
def increment_count(current_count):
    return current_count + 1

count = 10
count = increment_count(count)

print(count)
```

## Useful scope tricks

These patterns are handy when reading or debugging Python code:

- Use `locals()` to inspect variables available inside the current local scope.
- Use `globals()` carefully to inspect module-level names.
- Treat global constants as read-only values inside functions.
- Use `nonlocal` when an inner function must update a variable from an enclosing function.

```python
APP_NAME = "Python Training"

def show_scope_info(topic):
    level = "beginner"
    print(locals())
    print(globals()["APP_NAME"])  # Read a specific global name.

show_scope_info("scope")
```

### Updating an enclosing scope with `nonlocal`

`nonlocal` is used inside nested functions. It lets the inner function update a variable from the nearest enclosing function scope.

```python
def make_counter():
    count = 0

    def increment():
        nonlocal count
        count += 1
        return count

    return increment

counter = make_counter()

print(counter())
print(counter())
print(counter())
```
