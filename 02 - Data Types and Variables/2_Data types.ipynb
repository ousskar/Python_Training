# Data Types in Python

Python has several built-in data types. The most important question is not only *what values can this type hold?* but also *is it ordered, mutable, and how is it commonly used?*

## Quick map

- **Numeric types**: `int`, `float`, `complex`
- **Boolean type**: `bool`
- **Sequence types**: `str`, `list`, `tuple`, `range`
- **Set types**: `set`, `frozenset`
- **Mapping type**: `dict`
- **Special null value**: `None`

## Type checking and conversion

```python
print(type(10))          # <class 'int'>
print(type(3.14))        # <class 'float'>
print(type("Hello"))     # <class 'str'>

x = int("10")           # str -> int
y = float(10)           # int -> float
z = str(100)            # int -> str
```

As you read the examples below, pay attention to these questions:

- Is the type **ordered**?
- Can it be **modified in place**?
- Does it allow **duplicate values**?
- How is it usually used in real code?

````python
# Complex numbers have a real part and an imaginary part.
x = 3 + 4j
print(x)
print(type(x))
````

## 1. Numeric Types

Python numeric types let you represent whole numbers, decimal values, and complex numbers.

- `int` is used for whole numbers such as `10` or `-3`
- `float` is used for decimal numbers such as `3.14`
- `complex` is used when a number has a real and imaginary part such as `3 + 4j`

In day-to-day Python code, `int` and `float` are the most common numeric types.

````python
x = 10
pi = 3.14
z = 3 + 4j

print(x, type(x))
print(pi, type(pi))
print(z, type(z))

print(x + 5)
print(pi * 2)
print(z.real, z.imag)
````

## 2. Boolean Type

A boolean has only two possible values: `True` and `False`. Booleans are often produced by comparisons and are heavily used in `if` statements and loops.

````python
is_valid = True
is_equal = 10 == 10
is_greater = 5 > 8

print(is_valid, type(is_valid))
print(is_equal)
print(is_greater)
````

## 3. Sequence Types

Sequence types keep values in a specific order. Common sequence types include strings, lists, tuples, and ranges.

- `str` stores text
- `list` is ordered and **mutable**
- `tuple` is ordered and **immutable**
- `range` is useful for generating a sequence of integers

In the next examples:

- the list is modified after creation
- the tuple stays fixed unless we convert it to a list first

````python
name = "Python"
numbers = [1, 2, 3]          # list: ordered and mutable
coordinates = (10, 20)       # tuple: ordered and immutable
values = range(2, 5)         # 2, 3, 4

print(name)
print(numbers)
print(coordinates)
print(list(values))
````

````python
numbers.append(5)
numbers[0] = 100

print(numbers)
````

````python
print(coordinates)
print(type(coordinates))
````

````python
# Tuples do not support item assignment or append.

# coordinates.append(30)   # would raise AttributeError
````

````python
temp = list(coordinates)
temp.append(30)
updated_coordinates = tuple(temp)

print(temp)
print(updated_coordinates)
````

````python
temp
````

````python
for index, value in enumerate(temp):
    print(f"index={index}, value={value}")
````

### When should you use a tuple?

- Use a tuple when the data **should not be changed**.
- Use a tuple when you want a lightweight fixed collection.
- Use a tuple when you need a **hashable** sequence, for example as a dictionary key.
- Use a tuple when a function returns multiple values as one grouped result.

## 4. Set Types

A set stores **unique** values. Sets are useful when membership and uniqueness matter more than position.

- `set` is mutable
- `frozenset` is immutable

In the next examples, we inspect a set, add new values, and safely ignore a missing value with `discard()`.

````python
nums = {1, 2, 3}
fs = frozenset([1, 2, 3])

print(nums)
print(fs)
````

````python
print(type(fs))
print(list(fs))
````

````python
nums.discard(5)   # discard does not fail if the element is missing
print(nums)
````

````python
nums.add(2)
nums.add(20)
nums.add(100)
nums.add(10)

print(nums)
````

````python
nums.add(30)
nums.add(80)

print(nums)
````

````python
for index, value in enumerate(nums):
    print(f"index={index}, value={value}")

print("Set order should not be relied on.")
````

## 5. Dictionary Type

A dictionary stores data as **key-value pairs**. It is a good choice when you want to look up information by a meaningful key rather than by numeric position.

````python
country_capitals = {
    "Tunisia": "Tunis",
    "Germany": "Berlin",
    "Canada": "Ottawa",
    "England": "London",
}

print(country_capitals)
print(country_capitals["Tunisia"])
````

````python
### Common dictionary operations

- Access a value with a key: `country_capitals["Tunisia"]`
- Safely read a value: `country_capitals.get("Spain", "Not found")`
- Add or update: `country_capitals["Spain"] = "Madrid"`
- Iterate with `.items()` when you need both key and value
````

````python
for country, capital in country_capitals.items():
    print(f"{country} -> {capital}")

print(country_capitals.get("Spain", "Not found"))
country_capitals["Spain"] = "Madrid"
print(country_capitals)
````

## Quick comparison

- Use a **list** when order matters and values may change.
- Use a **tuple** when order matters and the data should stay fixed.
- Use a **set** when uniqueness matters more than order.
- Use a **dictionary** when you need to map keys to values.
- Use `None` when a variable intentionally has no value yet.

## Useful Python tricks with data types

These are small patterns that show up often in real Python code. They connect the basic data types to practical usage.

### 1. Truthiness

Many Python values can be used directly in conditions. Empty collections, `0`, and `None` are considered falsey, while non-empty values are truthy.

````python
print(bool(0))
print(bool(""))
print(bool([]))
print(bool([1, 2]))
print(bool(None))
````

### 2. Shared reference vs copy

Assigning one list to another variable does **not** create a new list. Use `.copy()` when you want a separate object.

````python
a = [1, 2]
b = a
c = a.copy()
b.append(3)

print("shared:", a, b)
print("copied:", c)
````

### 3. Deduplicating data with a set

A set removes duplicates automatically. This is a quick way to clean repeated values when order is not important.

````python
values = [1, 1, 2, 3, 3, 4]
unique_values = set(values)

print(values)
print(unique_values)
````

### 4. Indexing, slicing, and `isinstance()`

Slicing lets you extract part of a sequence, and `isinstance()` is a useful way to check whether a value belongs to a given type.

````python
text = "Python"
numbers = [10, 20, 30, 40]

print(text[0])
print(text[1:4])
print(numbers[:2])
print(isinstance(text, str))
print(isinstance(numbers, list))
````

## Advanced tricks with data types

These patterns are a step beyond the basics. They are still practical, but they introduce habits that become more useful in larger programs.

### 1. Nested copies and `deepcopy()`

A shallow copy duplicates only the outer container. If the structure contains inner lists or dictionaries, those inner objects are still shared.

````python
import copy

matrix = [[1, 2], [3, 4]]
shallow = matrix.copy()
deep = copy.deepcopy(matrix)

shallow[0][0] = 99

print("original:", matrix)
print("shallow:", shallow)
print("deep:", deep)
````

### 2. Dictionary merging

Python can merge dictionaries with the `|` operator. This is useful when combining configuration or lookup data.

````python
defaults = {"host": "localhost", "port": 8000}
override = {"port": 8080}
config = defaults | override

print(config)
````

### 3. Set operations

Sets are not only for removing duplicates. They are also useful for comparing groups of values with intersection, union, and difference.

````python
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(a | b)   # union
print(a & b)   # intersection
print(a - b)   # difference
````

### 4. Slicing with a step

A slice can also use a third value called the step. This lets you skip values or reverse a sequence.

````python
text = "abcdefgh"

print(text[::2])
print(text[::-1])
````

### 5. Safe numeric conversion

When converting user input, `float()` can accept values like `"10.5"` that `int()` cannot parse directly.

````python
print(int("10"))
print(float("10.5"))

# print(int("10.5"))  # would raise ValueError
````
