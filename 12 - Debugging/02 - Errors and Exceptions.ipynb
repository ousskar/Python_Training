# Errors and Exceptions

In Python, errors and exceptions can interrupt the execution of program.

Python provides try and except blocks to handle situations like this. In case an error occurs in try-block, Python stops executing try block and jumps to exception block. These blocks let you handle the errors without crashing the program.



```
Syntax:
  try:
    # Code that might raise an exception
    risky_operation()
except ExceptionType:
    # Code to handle the exception
    handle_error()
```



# Common Examples

Handling a specific exception

````python
try:
    result = 10 / 0
except ZeroDivisionError:
    print("You can't divide by zero!")
````

Handling multiple exceptions

````python
try:
    # Code that might raise different exceptions
    value = int("abc")
    result = 10 / value
except ValueError:
    print("That's not a valid number!")
except ZeroDivisionError:
    print("You can't divide by zero!")
````

Catching all exceptions (generally not recommended)

````python
try:
    # Risky code
    undefined_function()
except:
    print("Something went wrong!")
````

Using else (executes if no exception occurs)

````python
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Division by zero!")
else:
    print("Result is:", result)  # Only runs if no exception
````

Using finally (always executes)

````python
def divide_numbers(a, b):
    try:
        result = a / b
        print(f"Result: {result}")
    except ZeroDivisionError:
        print("Error: Cannot divide by zero.")
    finally:
        print("Execution of try-except block is complete.")

# Example usage
divide_numbers(10, 2)   # Normal case
print("-----")
divide_numbers(5, 0)    # Will raise exception
````

## Common Exceptions You Might Use
-----------------------------------------
ValueError :	Invalid value (e.g. int("abc"))

TypeError : Wrong type (e.g. adding string to int)

IndexError :	List index out of range

KeyError :	Missing key in a dictionary

ZeroDivisionError :	Dividing by zero

FileNotFoundError :	File or path doesn't exist

AttributeError :	Accessing non-existent attribute

ImportError :	Module not found

-------------------------------------

If you're unsure which error might occur: Temporarily use except Exception as e for debugging

````python
try:
    some_risky_code()
except Exception as e:
    print("Caught error:", type(e).__name__, "-", str(e))
````

# Summary:
*  The try block lets you test a block of code for errors.
*  The except block lets you handle the error.
*  The else block lets you execute code when there is no error.
*  The finally block lets you execute code, regardless of the result of the try- and except blocks.
