# 5. LLM Capstone Project

This capstone asks you to build a complete Python project that uses local study material and LLM-style features to create a useful learning assistant.

The goal is not to train a language model. The goal is to build a Python application that uses everything you learned in this course to power an LLM-based study workflow.

## 1. Project Title

**Build an LLM Study Assistant in Python**

## 2. Project Goal

Create a Python application called `StudyMate LLM` that helps a learner work with local course materials.

The application should:

- load study notes from `.txt`, `.csv`, or `.json` files
- clean and organize the content
- split large content into chunks
- generate summaries
- answer study questions
- generate flashcards or quizzes
- save study sessions for later review

## 3. Why This Project Fits the Course

This capstone uses nearly every topic from the course:

- Python basics for data handling and control flow
- iterables, indexing, and slicing for chunking and batching
- functions for reusable program logic
- files for loading and saving data
- modules and packages for code organization
- OOP for modeling app components
- decorators for logging or timing
- testing for core logic
- debugging for error handling and improvement

## 4. Minimum Requirements

Your project must include:

- at least 5 reusable functions
- at least 1 class
- reading from at least 2 file formats
- saving output to JSON or text
- exception handling for invalid files or bad input
- at least 1 decorator
- automated tests for key logic
- at least 1 mocked or fake LLM interaction in tests

## 5. Required Features

Your app should include at least these commands or features:

1. Load content
2. Summarize notes
3. Ask a question about the notes
4. Generate flashcards
5. Generate a short quiz
6. Save the current session

## 6. Suggested Project Structure

```text
studymate_llm/
  app.py
  loader.py
  cleaner.py
  chunker.py
  prompts.py
  llm_client.py
  quiz.py
  storage.py
  models.py
  utils.py
tests/
  test_loader.py
  test_chunker.py
  test_quiz.py
  test_storage.py
data/
  sample_notes.txt
  sample_terms.csv
sessions/
README.md
```

## 7. Milestones

### Milestone 1 – Core Setup
- create the folder structure
- build a command-line menu
- load one text file

### Milestone 2 – Data Preparation
- load CSV or JSON content
- clean whitespace and empty values
- save cleaned data

### Milestone 3 – Content Processing
- split long text into chunks
- create prompt-building functions
- organize data for summary and quiz generation

### Milestone 4 – LLM Layer
- create an `LLMClient` class
- start with a fake response generator
- optionally swap in a real API later

### Milestone 5 – Study Features
- generate summaries
- answer questions
- generate flashcards and quizzes

### Milestone 6 – Testing and Debugging
- add tests for loaders, chunking, and quiz logic
- mock the LLM layer
- debug file and content edge cases

## 8. Stretch Goals

If you finish early, try one or more of these:

- add difficulty levels for quiz generation
- store flashcards in a reusable JSON format
- add a simple retrieval step that picks the most relevant chunk before asking a question
- add logging and execution timing decorators
- track prompt and response metadata
