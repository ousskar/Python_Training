# Python Debugging Tools

In this notebook, you will use Python's dedicated debugging tools to pause execution, inspect variables, and step through code more deliberately than with `print()` alone.

This notebook covers:

- `pdb` and `breakpoint()`
- stepping through code
- inspecting variables while a program is paused
- using IDE debuggers such as VS Code

## 1. Using `pdb`

`pdb` is Python's built-in interactive debugger. You can pause execution at a chosen line, inspect variables, and continue one step at a time.

Common commands:

- `l` to list nearby code
- `n` to run the next line
- `s` to step into a function call
- `p expression` to print a value
- `c` to continue execution
- `q` to quit the debugger

```python
import pdb

def buggy_total(prices, discount):
    subtotal = sum(prices)
    pdb.set_trace()
    final_total = subtotal - discount
    return final_total

buggy_total([19.99, 5.00, 3.50], "2")
```

## 2. Using `breakpoint()`

In Python 3.7 and later, `breakpoint()` is a convenient shortcut that starts the debugger without importing `pdb` explicitly.

```python
def inspect_discount(prices, discount):
    subtotal = sum(prices)
    breakpoint()
    return subtotal - discount

inspect_discount([10, 20, 30], 5)
```

## 3. Inspecting Variables While Paused

When execution stops at a breakpoint, inspect the values that control the next step in your program.

Useful checks include:

- the type of each variable
- whether a value is unexpectedly `None`
- whether a list or dictionary contains the data you expect

```python
def inspect_cart(cart):
    total_items = len(cart)
    breakpoint()
    total_price = sum(item["price"] for item in cart)
    return total_items, total_price

inspect_cart([
    {"name": "Notebook", "price": 12.5},
    {"name": "Pen", "price": 1.5},
])
```

## 4. IDE Debuggers

Modern editors can wrap the same ideas in a visual interface. In VS Code or PyCharm, you typically:

1. set a breakpoint in the gutter
2. start the program in debug mode
3. use step controls to move through execution
4. inspect variables, the call stack, and watched expressions

## 5. Logging and Assertions

`print()` is useful while learning, but larger programs benefit from logging and assertions:

- logging helps you capture program state without editing every call site later
- assertions help you fail early when a condition that should always be true becomes false

```python
import logging

logging.basicConfig(level=logging.INFO, format="%(levelname)s: %(message)s")

def process_order(total, tax_rate):
    assert tax_rate >= 0, "tax_rate must not be negative"
    logging.info("Starting order processing")
    logging.info("Total before tax: %s", total)
    return total * (1 + tax_rate)

process_order(100, 0.2)
```
