## Exercise 1 – Basic Dataclass

**Task**

Create a dataclass `Book` with:
- `title: str`
- `author: str`
- `price: float`
- `in_stock: bool` (default `True`)

Then:
1. Create 3 book instances.
2. Put them in a list.
3. Print the list and observe the auto `__repr__`.


````python
from dataclasses import dataclass

@dataclass
class Book:
    title: str
    author: str
    price: float
    in_stock: bool = True

# Create 3 book instances
book1 = Book("Clean Code", "Robert C. Martin", 35.5)
book2 = Book("Fluent Python", "Luciano Ramalho", 45.0, in_stock=False)
book3 = Book("Python Tricks", "Dan Bader", 25.0)

books = [book1, book2, book3]
books  # Notice the nice auto-generated representation for each Book
````

## Exercise 2 – Classroom with Students

**Task**

Create:
```python
@dataclass
class Student:
    name: str
    grade: float

@dataclass
class Classroom:
    name: str
    students: list[Student] = field(default_factory=list)
```

Then:
1. Add a method `add_student(self, student: Student)` to `Classroom`.
2. Add a method `average_grade(self) -> float` that returns the average grade (0 if no students).
3. Test with several students.


````python
from dataclasses import dataclass, field
from typing import List

@dataclass
class Student:
    name: str
    grade: float

@dataclass
class Classroom:
    name: str
    students: List[Student] = field(default_factory=list)

    def add_student(self, student: Student) -> None:
        """Add a student to the classroom."""
        self.students.append(student)

    def average_grade(self) -> float:
        """Return the average grade of students, or 0.0 if there are none."""
        if not self.students:
            return 0.0
        total = sum(s.grade for s in self.students)
        return total / len(self.students)

# Test the implementation
c = Classroom("Dataclasses 101")
c.add_student(Student("Alice", 18.0))
c.add_student(Student("Bob", 14.0))
c.add_student(Student("Clara", 16.0))

print("Classroom:", c)
print("Average grade:", c.average_grade())
````

## Exercise 3 – Validation with `__post_init__`

**Task**

Extend `Student` so that:
- Grade must be between 0 and 20.
- If not, raise `ValueError` in `__post_init__`.

Test by creating a `Student` with invalid grade and observe the error.


````python
from dataclasses import dataclass

@dataclass
class ValidatedStudent:
    name: str
    grade: float

    def __post_init__(self):
        if not (0 <= self.grade <= 20):
            raise ValueError("Grade must be between 0 and 20")

# Valid example
vs1 = ValidatedStudent("Alice", 15.0)
print(vs1)

# Invalid example (uncomment to see the error)
# vs2 = ValidatedStudent("Bob", 25.0)
````

## Exercise 4 – Immutable Configuration

**Task**

Create a frozen dataclass:
```python
@dataclass(frozen=True)
class Config:
    host: str
    port: int
```

Try to modify `host` after creation and observe the error.


````python
from dataclasses import dataclass

@dataclass(frozen=True)
class Config:
    host: str
    port: int

cfg = Config("localhost", 8080)
print(cfg)

# Attempting to modify will raise dataclasses.FrozenInstanceError
try:
    cfg.host = "127.0.0.1"  # type: ignore[attr-defined]
except Exception as e:
    print("Error when trying to modify frozen Config:", e)
````

## Exercise 5 – Sorting Players

**Task**

Create:
```python
@dataclass(order=True)
class Player:
    score: int
    name: str
```

1. Create several players with different scores.
2. Sort them and print them.
3. (Optional) Add a field `level` that is **not** used for comparison (use `compare=False`).


````python
from dataclasses import dataclass, field

@dataclass(order=True)
class Player:
    # The first field(s) define the primary sort keys
    score: int
    name: str
    # Optional field not used for comparison
    level: int = field(default=1, compare=False)

players = [
    Player(10, "Alice", level=3),
    Player(5, "Bob", level=1),
    Player(20, "Clara", level=5),
    Player(10, "Dave", level=2),
]

print("Original order:")
for p in players:
    print(p)

print("\nSorted players (by score, then name):")
for p in sorted(players):
    print(p)
````
