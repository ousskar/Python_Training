## 3. Temporary Files with `tmp_path`

Pytest provides the built-in `tmp_path` fixture when a test needs a temporary directory. This is useful for file-based tests that should not touch real project files.

```python
def test_needs_files(tmp_path):
    example_file = tmp_path / "example.txt"
    example_file.write_text("hello")

    assert example_file.exists()
    assert example_file.read_text() == "hello"
```

# Skipping test functions

```python
@pytest.mark.skip(reason="no way of currently testing this")
def test_the_unknown(): ...
```

```python
def valid_config():
    return False

def test_function():
    if not valid_config():
        pytest.skip("unsupported configuration")
```

```python
# It is also possible to skip the whole module using pytest.skip(reason, allow_module_level=True) at the module level:
import sys

import pytest

if not sys.platform.startswith("win"):
    pytest.skip("skipping windows-only tests", allow_module_level=True)
```

# skipif
If you wish to skip something conditionally then you can use skipif instead.

```python
import sys


@pytest.mark.skipif(sys.version_info < (3, 10), reason="requires python3.10 or higher")
def test_function(): ...
```

# Skip all test functions of a class or module


```python
@pytest.mark.skipif(sys.platform == "win32", reason="does not run on windows")
class TestPosixCalls:
    def test_function(self):
        "will not be setup or run under 'win32' platform"
```

# XFail: mark test functions as expected to fail

You can use the xfail marker to indicate that you expect a test to fail.

```python
@pytest.mark.xfail
def test_function(): ...
```

This test will run but no traceback will be reported when it fails. Instead, terminal reporting will list it in the “expected to fail” (XFAIL) or “unexpectedly passing” (XPASS) sections.

Alternatively, you can also mark a test as XFAIL from within the test or its setup function imperatively:


```python
def valid_config():
    return False

def test_function():
    if not valid_config():
        pytest.xfail("failing configuration (but should work)")
```

Example:

```python
from __future__ import annotations

import pytest


xfail = pytest.mark.xfail


@xfail
def test_hello():
    assert 0


@xfail(run=False)
def test_hello2():
    assert 0


@xfail("hasattr(os, 'sep')")
def test_hello3():
    assert 0


@xfail(reason="bug 110")
def test_hello4():
    assert 0


@xfail('pytest.__version__[0] != "17"')
def test_hello5():
    assert 0


def test_hello6():
    pytest.xfail("reason")


@xfail(raises=IndexError)
def test_hello7():
    x = []
    x[1] = 1
```

# Skip/xfail with parametrize

```python
import sys

import pytest


@pytest.mark.parametrize(
    ("n", "expected"),
    [
        (1, 2),
        pytest.param(1, 0, marks=pytest.mark.xfail),
        pytest.param(1, 3, marks=pytest.mark.xfail(reason="some bug")),
        (2, 3),
        (3, 4),
        (4, 5),
        pytest.param(
            10, 11, marks=pytest.mark.skipif(sys.version_info >= (3, 0), reason="py2k")
        ),
    ],
)
def test_increment(n, expected):
    assert n + 1 == expected
```

## 4. Pytest Fixtures

Fixtures are reusable setup functions that provide data or services to tests. They help reduce duplication and make tests easier to read and maintain.

```python
import pytest

@pytest.fixture
def input_value():
   input = 39
   return input

def test_divisible_by_3(input_value):
   assert input_value % 3 == 0

def test_divisible_by_6(input_value):
   assert input_value % 6 == 0
```
