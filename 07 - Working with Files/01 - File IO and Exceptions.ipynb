# File I/O and Exceptions in Python

This notebook introduces practical file handling in Python. You will create file paths, write text, read text, append content, use `with` for safe file handling, copy binary data, and handle common file errors.

The examples are self-contained and write to a small `file_io_examples` folder next to this notebook.

## 1. Data persistence and file paths

Data persistence means saving data outside program memory so it is still available after the program stops.

Python can work with plain strings as paths, but `pathlib.Path` is usually clearer and safer because it joins path parts correctly on different operating systems.

## 2. Creating paths with `pathlib`

A path points to a file or folder. The `/` operator joins path parts when using `Path` objects.

`mkdir(exist_ok=True)` creates a folder if it does not already exist and does nothing if it is already there.

```python
from pathlib import Path

example_dir = Path("file_io_examples")
example_dir.mkdir(exist_ok=True)

text_path = example_dir / "notes.txt"
log_path = example_dir / "activity.log"
binary_path = example_dir / "sample.bin"
binary_copy_path = example_dir / "sample_copy.bin"

print(f"Example folder: {example_dir.resolve()}")
```

## 3. Opening and closing files

Python's built-in `open()` function opens a file and returns a file object. A file should be closed when you are done with it.

Common modes:

- `'r'`: read
- `'w'`: write and replace existing content
- `'a'`: append to the end
- `'x'`: create a new file and fail if it already exists
- `'b'`: binary mode
- `'t'`: text mode, the default

For most text files, also pass `encoding="utf-8"`.

```python
file = open(text_path, "w", encoding="utf-8")
file.write("Python file handling\n")
file.write("Write, read, append, and close files.\n")
file.close()

print(f"Created: {text_path}")
```

## 4. Reading text files

You can read the entire file at once with `.read()`. This is simple and useful for small files.

When you read the file later, you get back one string.

```python
file = open(text_path, "r", encoding="utf-8")
content = file.read()
file.close()

print(content)
```

### Reading line by line

For larger files, read one line at a time instead of loading everything into memory.

Use `.strip()` to remove the newline character from the end of each line before printing.

```python
file = open(text_path, "r", encoding="utf-8")

for line_number, line in enumerate(file, start=1):
    clean_line = line.strip()
    print(f"{line_number}: {clean_line}")

file.close()
```

## 5. Writing and appending to files

Mode `'w'` replaces existing content. Mode `'a'` appends new content to the end of the file.

Remember to include `\n` when you want the next piece of text to start on a new line.

```python
file = open(text_path, "a", encoding="utf-8")
file.write("Appended line\n")
file.close()

file = open(text_path, "r", encoding="utf-8")
updated_content = file.read()
file.close()

print(updated_content)
```

## 6. Context managers with `with`

The `with` statement closes the file automatically, even if an error happens inside the block.

This is the preferred pattern for most file work because it keeps file handling short, readable, and safer.

```python
with open(log_path, "w", encoding="utf-8") as log_file:
    log_file.write("Program started\n")
    log_file.write("Program finished\n")

with open(log_path, "r", encoding="utf-8") as log_file:
    print(log_file.read())
```

## 7. Useful path checks

Before reading or writing, it is often useful to ask simple questions about a path:

- Does it exist?
- Is it a file?
- Is it a folder?
- What is its absolute location?

These checks are useful for debugging file problems.

```python
print(f"Exists: {text_path.exists()}")
print(f"Is file: {text_path.is_file()}")
print(f"Is folder: {text_path.is_dir()}")
print(f"Absolute path: {text_path.resolve()}")
```

## 8. Exception handling

File operations can fail when a path does not exist, permission is denied, or data cannot be decoded.

Use `try` / `except` when you expect a recoverable error and can respond clearly.

```python
missing_path = example_dir / "missing.txt"

try:
    with open(missing_path, "r", encoding="utf-8") as file:
        print(file.read())
except FileNotFoundError:
    print(f"Could not find: {missing_path}")
except OSError as error:
    print(f"Could not read the file: {error}")
```

## 9. Binary files

Text mode is for strings. Binary mode is for bytes.

Use binary mode for images, PDFs, audio files, compressed files, and any file where Python should not try to decode the content as text.

```python
binary_path.write_bytes(bytes(range(10)))

with open(binary_path, "rb") as source_file, open(binary_copy_path, "wb") as target_file:
    while chunk := source_file.read(4):
        target_file.write(chunk)

print(binary_path.read_bytes())
print(binary_copy_path.read_bytes())
print(binary_path.read_bytes() == binary_copy_path.read_bytes())
```
