# `*args` and `**kwargs`

Python functions can receive arguments in two main ways:

- **Positional arguments** are matched by position: `greet("Alice")`.
- **Keyword arguments** are matched by name: `greet(name="Alice")`.

Use `*args` when a function should accept any number of extra positional arguments. Inside the function, `args` behaves like a tuple.

````python
def sum_all_numbers(*numbers):
    total = 0

    for number in numbers:
        total += number

    return total

print(sum_all_numbers(1, 2, 3))
print(sum_all_numbers(10, 20, 30, 40, 50))
````

## `**kwargs`

Use `**kwargs` when a function should accept any number of extra keyword arguments. Inside the function, `kwargs` behaves like a dictionary.

````python
def display_record(**fields):
    for key, value in fields.items():
        print(f"{key}: {value}")

display_record(name="Alice", age=30, city="New York")
print("\n---")
display_record(product="Laptop", price=1200, brand="TechCo")
````

## Combining regular arguments, `*args`, and `**kwargs`

The usual order in a function definition is:

```python
def function(required, default_value=10, *args, keyword_only=True, **kwargs):
    ...
```

In simple terms: regular positional parameters come first, `*args` collects extra positional values, keyword-only parameters can come after `*args`, and `**kwargs` comes last.

### Example: mixed argument types

The next function receives one required value, extra positional values, a keyword-only option, and extra keyword metadata.

````python
def build_report(title, *sections, uppercase_title=False, **metadata):
    if uppercase_title:
        title = title.upper()

    print(f"Report: {title}")

    for section in sections:
        print(f"- {section}")

    if metadata:
        print("Metadata:")
        for key, value in metadata.items():
            print(f"  {key}: {value}")

build_report(
    "Weekly progress",
    "Functions completed",
    "Scope reviewed",
    uppercase_title=True,
    author="Alice",
    version=1,
)
````

## Unpacking arguments when calling a function

`*` and `**` can also be used when calling a function:

- `*values` unpacks a list or tuple into positional arguments.
- `**options` unpacks a dictionary into keyword arguments.

This is useful when arguments are stored in collections.

````python
def describe_lesson(title, duration, level="beginner"):
    print(f"{title} | {duration} minutes | {level}")

lesson_values = ["Args and Kwargs", 45]
lesson_options = {"level": "intermediate"}

describe_lesson(*lesson_values, **lesson_options)
````

## Advanced preview: forwarding arguments

Sometimes a function receives flexible arguments and passes them to another function. This is called forwarding arguments. You will see this pattern again in decorators and classes.

````python
def log_message(message, level="INFO", user="system"):
    print(f"[{level}] ({user}) {message}")

def log_event(message, *args, **kwargs):
    print("Preparing log event...")
    log_message(message, *args, **kwargs)

log_event("Notebook opened")
log_event("Function lesson updated", level="DEBUG", user="trainer")
````
