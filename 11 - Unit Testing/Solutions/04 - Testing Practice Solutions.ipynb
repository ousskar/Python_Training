## 1. Practice Function

```python
def normalize_username(name):
    if not isinstance(name, str):
        raise TypeError("name must be a string")

    cleaned = name.strip().lower()
    if not cleaned:
        raise ValueError("name cannot be empty")

    return cleaned.replace(" ", "_")
```

## 2. `unittest` Test Cases Solution

```python
import unittest

class TestNormalizeUsername(unittest.TestCase):
    def test_normalizes_case_and_spaces(self):
        self.assertEqual(normalize_username("  Alice Smith  "), "alice_smith")

    def test_rejects_empty_string(self):
        with self.assertRaises(ValueError):
            normalize_username("   ")

    def test_rejects_non_string_input(self):
        with self.assertRaises(TypeError):
            normalize_username(123)

unittest.main(argv=['first-arg-is-ignored'], exit=False)
```

## 3. Improved Discount Tests

```python
def apply_discount(price, percent):
    if percent < 0 or percent > 100:
        raise ValueError("percent must be between 0 and 100")
    return price - (price * percent / 100)

def test_apply_discount():
    assert apply_discount(200, 10) == 180

def test_apply_discount_invalid_percent():
    try:
        apply_discount(200, 120)
    except ValueError:
        pass
    else:
        raise AssertionError("Expected ValueError for invalid percent")
```

## 4. Mocking Exercise Solution

```python
from unittest.mock import Mock

def notify_user(user_email, sender):
    sender.send(user_email, "Your report is ready")
    return True

fake_sender = Mock()
result = notify_user("user@example.com", fake_sender)

print(result)
fake_sender.send.assert_called_once_with("user@example.com", "Your report is ready")
```

## 5. Challenge Exercise Solution

The bug is that `last_name` incorrectly uses the first element instead of the last element.

```python
def split_full_name(full_name):
    parts = full_name.split()
    if len(parts) < 2:
        raise ValueError("full_name must contain at least two words")
    first_name = parts[0]
    last_name = parts[-1]
    return first_name, last_name

def test_split_full_name_two_words():
    assert split_full_name("Alice Smith") == ("Alice", "Smith")

def test_split_full_name_extra_spaces():
    assert split_full_name("  Bob   Jones  ") == ("Bob", "Jones")

def test_split_full_name_multi_part_name():
    assert split_full_name("Mary Jane Watson") == ("Mary", "Watson")

def test_split_full_name_invalid_input():
    try:
        split_full_name("Prince")
    except ValueError:
        pass
    else:
        raise AssertionError("Expected ValueError for a one-word name")
```
