# Object Serialization and Deserialization in Python

This notebook introduces **JSON**, the process of **serialization** and **deserialization**, package management with **pip**, and basic **Web API** access using Python.

## 1. Why Serialization?

**Serialization** converts Python objects into a storable or transmittable format. **JSON** is the most widely used format for this purpose because it is lightweight and text-based.



## 2. JSON Overview

JSON (JavaScript Object Notation) structure looks very similar to Python dictionaries and lists. Below is an example with English labels:

```python
{
  "name": "John Smith",
  "age": 40,
  "friends": ["William", "Alice"],
  "cars": [
    {"brand": "BMW", "max_speed": 240},
    {"brand": "Benz", "max_speed": 280},
    {"brand": "Audi", "max_speed": 280}
  ]
}
```

## 3. JSON ↔ Python Type Mapping

| JSON | Python |
|------|--------|
| object | dict |
| array | list |
| string | str |
| number | int / float |
| boolean | bool |
| null | None |

## 4. Serializing Python Objects (`json.dumps`)

Use `json.dumps()` to turn a dictionary into a JSON-formatted string.

```python
import json

user_dict = {
    'name': 'John Smith',
    'age': 40,
    'friends': ['William', 'Alice'],
    'cars': [
        {'brand': 'BMW', 'max_speed': 240},
        {'brand': 'Audi', 'max_speed': 280},
        {'brand': 'Benz', 'max_speed': 280}
    ]
}

json_str = json.dumps(user_dict, indent=2)
print(json_str)
```

## 5. File Operations: `dump` and `load`

```python
# Writing JSON to a file
with open('data.json', 'w', encoding='utf-8') as f:
    json.dump(user_dict, f, indent=2)

# Reading JSON from a file
with open('data.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
    print(data)
```
