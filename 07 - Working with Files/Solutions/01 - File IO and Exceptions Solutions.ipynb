## Practice tasks

1. Create a file named `shopping_list.txt` inside `file_io_examples`.
2. Write three shopping items to the file, one item per line.
3. Read the file and print each item with its line number.
4. Append one more item to the file.
5. Try to read a file that does not exist and handle the error with a friendly message.

```python
from pathlib import Path

# Use the shared examples folder next to the lesson notebook
example_dir = Path("..") / "file_io_examples"
example_dir.mkdir(exist_ok=True)

shopping_list = example_dir / "shopping_list.txt"

# 1–2. Create the file and write three items (one per line)
shopping_list.write_text("milk\nbread\neggs\n", encoding="utf-8")
print(f"Created: {shopping_list.resolve()}")
```

```python
# 3. Read the file and print each item with its line number
with shopping_list.open("r", encoding="utf-8") as f:
    for i, line in enumerate(f, start=1):
        print(f"{i}. {line.strip()}")
```

```python
# 4. Append one more item
with shopping_list.open("a", encoding="utf-8") as f:
    f.write("butter\n")

print("After append:")
print(shopping_list.read_text(encoding="utf-8"))
```

```python
# 5. Handle a missing file with a friendly message
missing = example_dir / "does_not_exist.txt"

try:
    text = missing.read_text(encoding="utf-8")
    print(text)
except FileNotFoundError:
    print(f"Could not find file: {missing.name}")
    print("Check the file name and try again.")
```

## Alternative: all steps in one cell

A compact version of the same solution:

```python
from pathlib import Path

example_dir = Path("..") / "file_io_examples"
example_dir.mkdir(exist_ok=True)
shopping_list = example_dir / "shopping_list.txt"

shopping_list.write_text("milk\nbread\neggs\n", encoding="utf-8")

with shopping_list.open("r", encoding="utf-8") as f:
    for i, line in enumerate(f, start=1):
        print(f"{i}. {line.strip()}")

with shopping_list.open("a", encoding="utf-8") as f:
    f.write("butter\n")

try:
    Path(example_dir / "does_not_exist.txt").read_text(encoding="utf-8")
except FileNotFoundError:
    print("Could not find file: does_not_exist.txt")
```
