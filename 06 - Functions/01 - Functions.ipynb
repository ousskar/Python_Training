# Functions

Functions let you give a name to a reusable block of code. They help you avoid repetition, make programs easier to read, and break larger problems into smaller steps.

## Why functions matter

When you notice yourself writing the same logic more than once, that is often a sign that a function would help. Start by writing the code once, then move the repeated idea into a function with a clear name.

````python
length = 10
width = 5

area = length * width
perimeter = 2 * (length + width)
diagonal = (length ** 2 + width ** 2) ** 0.5

print(f"Area: {area}")
print(f"Perimeter: {perimeter}")
print(f"Diagonal: {diagonal:.2f}")
````

````python
def rectangle_properties(length, width):
    area = length * width
    perimeter = 2 * (length + width)
    diagonal = (length ** 2 + width ** 2) ** 0.5
    return {
        "area": area,
        "perimeter": perimeter,
        "diagonal": diagonal,
    }

result = rectangle_properties(10, 5)

print(f"Area: {result['area']}")
print(f"Perimeter: {result['perimeter']}")
print(f"Diagonal: {result['diagonal']:.2f}")
````

## Function anatomy

A basic function has a name, parameters, a body, and often a `return` value.

```python
def function_name(parameter):
    result = parameter * 2
    return result
```

Parameters receive input. `return` sends a result back to the code that called the function.

````python
def rectangle_properties(length=10, width=5):
    """Return area, perimeter, and diagonal for a rectangle."""
    area = length * width
    perimeter = 2 * (length + width)
    diagonal = (length ** 2 + width ** 2) ** 0.5

    return {
        "area": area,
        "perimeter": perimeter,
        "diagonal": diagonal,
    }

default_rectangle = rectangle_properties()
custom_rectangle = rectangle_properties(length=8, width=3)

print(default_rectangle)
print(custom_rectangle)
````

## Getting help on functions

A docstring is not only a comment for people reading the source code. Python stores it on the function, so tools can display it later.

Useful ways to inspect a function:

- `help(function_name)` shows the function name and docstring.
- `function_name.__doc__` returns the docstring as a string.
- In VS Code, hovering over a function call can also show the docstring.

````python
def calculate_discount(price, discount_rate):
    """Return the price after applying a discount rate.

    price: original price before the discount
    discount_rate: decimal discount, such as 0.20 for 20%
    """
    return price * (1 - discount_rate)

#print(calculate_discount.__doc__)

help(calculate_discount)
````

## Copies and references in function arguments

When you pass a variable to a function, Python passes a reference to the object. This matters most for mutable objects such as lists and dictionaries.

- If a function mutates a list or dictionary, the original object can change too.
- Use `.copy()` when you want to protect the original top-level collection.
- Avoid mutable default parameters like `items=[]` because the same object is reused across calls.

````python
def add_tag(tags, new_tag):
    tags.append(new_tag)  # This mutates the original list.
    return tags

course_tags = ["python", "functions"]
updated_tags = add_tag(course_tags, "mutable")

print(updated_tags)
print(course_tags)
````

### Protecting the original with `.copy()`

If the function should return a modified version without changing the original list, make a copy inside the function.

````python
def add_tag_safely(tags, new_tag):
    copied_tags = tags.copy()
    copied_tags.append(new_tag)
    return copied_tags

course_tags = ["python", "functions"]
updated_tags = add_tag_safely(course_tags, "copy")

print(updated_tags)
print(course_tags)
````

### Avoid mutable default parameters

Default parameter values are created once, when the function is defined. If the default value is a list or dictionary, changes can unexpectedly appear in later calls.

````python
def add_item_bad(item, items=[]):
    items.append(item)
    return items

print(add_item_bad("first"))
print(add_item_bad("second"))  # The previous value is still there.

def add_item_good(item, items=None):
    if items is None:
        items = []

    items.append(item)
    return items

print(add_item_good("first"))
print(add_item_good("second"))
````

## Function error checking and handling

Functions should protect their assumptions. If a function needs a positive number, a non-empty list, or a specific type of value, check that early and raise a clear error.

A common pattern is:

1. Validate inputs at the top of the function.
2. Raise `ValueError` when the value has the right type but an invalid value.
3. Handle the error near the function call when you want the program to continue.

````python
def apply_discount(price, discount_rate):
    if price < 0:
        raise ValueError("price must be zero or greater")

    if not 0 <= discount_rate <= 1:
        raise ValueError("discount_rate must be between 0 and 1")

    return price * (1 - discount_rate)

print(apply_discount(100, 0.20))

try:
    print(apply_discount(100, 1.50))
except ValueError as error:
    print(f"Invalid input: {error}")
````

## Inner functions

An inner function is a function defined inside another function. Use it when a helper is only needed by one outer function and does not need to be available everywhere else.

Inner functions are useful for:

- organizing helper logic close to where it is used
- avoiding extra names in the global scope
- sharing values from the outer function

````python
def format_label(text, prefix):
    def clean(value):
        return value.strip().title()

    clean_text = clean(text)
    clean_prefix = clean(prefix)

    if not clean_text:
        raise ValueError("text cannot be empty")

    return f"{clean_prefix}: {clean_text}"

print(format_label("  python functions  ", "topic"))
print(format_label("  scope  ", "section"))
````

## Preview: generator functions

A normal function returns one result and then stops. A generator function uses `yield` to produce values one at a time. This is useful when a sequence could be large or when values are produced gradually.

````python
def first_numbers(limit):
    number = 0

    while number < limit:
        yield number
        number += 1

for value in first_numbers(5):
    print(value)
````

## Preview: flexible arguments

Sometimes a function needs to accept a flexible number of values. `*args` collects extra positional arguments into a tuple. `**kwargs` collects extra keyword arguments into a dictionary. A later notebook covers this in more detail.

````python
def add_all(*numbers):
    total = 0

    for number in numbers:
        total += number

    return total

print(add_all(1, 2, 3))
print(add_all(10, 20, 30, 40))
````

````python
def build_profile(**details):
    return ", ".join(f"{key}={value}" for key, value in details.items())

print(build_profile(name="Alice", role="trainer", language="Python"))
````

# Lambda functions

A `lambda` is a small anonymous function written in one expression. Use it for simple, short operations, especially as an argument to another function. Use `def` for logic that needs a clear name or multiple steps.

```python
lambda argument: expression
```

````python
square = lambda number: number ** 2

print(square(5))
````

````python
words = ["banana", "apple", "cherry", "date"]

sorted_words = sorted(words, key=lambda word: len(word))

print(sorted_words)
````

## Lambda limitations

A lambda can contain only one expression. If the logic needs comments, multiple steps, error handling, or a descriptive name, use `def` instead.

## Practice: write small functions

Try these prompts after the examples:

1. Write a function `greet(name)` that returns `"Hello, <name>"`.
2. Write a function `is_even(number)` that returns `True` for even numbers and `False` otherwise.
3. Write a function `clean_text(text)` that strips whitespace and lowercases the text.
4. Write a function `chunk_words(text, chunk_size)` that splits text into word chunks.

````python
def greet(name):
    return f"Hello, {name}"

def is_even(number):
    return number % 2 == 0

def clean_text(text):
    return text.strip().lower()

def chunk_words(text, chunk_size):
    words = text.split()
    chunks = []

    for start in range(0, len(words), chunk_size):
        chunk = words[start:start + chunk_size]
        chunks.append(" ".join(chunk))

    return chunks

print(greet("Alice"))
print(is_even(10))
print(is_even(7))
print(clean_text("  HELLO Python  "))
print(chunk_words("Functions help us reuse code in small clear pieces", 3))
````
