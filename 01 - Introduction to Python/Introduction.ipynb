# Introduction to Python

Python is a high-level programming language created by Guido van Rossum, a Dutch computer scientist. As shown in the January 2026 TIOBE Index, Python ranks number one worldwide, with a rating of 22.61%, confirming its position as the most popular programming language today, well ahead of C, Java, and C++.

Python’s success is largely due to its strong emphasis on readability, simplicity, and expressive power. Compared to influential languages such as C, C++, and Java, Python allows developers to implement ideas with significantly fewer lines of code, making it especially suitable for rapid development, education, and problem solving.

The ranking illustrated in the chart reflects real-world usage across industries, academia, and open-source communities. Python’s dominance is driven by its widespread adoption in data science, artificial intelligence, machine learning, web development, automation, and scientific computing. Its consistent leadership in authoritative indexes like TIOBE demonstrates not only its popularity but also its long-term relevance and stability as a core programming language.

Below are some important milestones in the development of the Python language:

1. December 1989: Guido van Rossum resolved to develop a new scripting language and its interpreter to pass the boring Christmas season. The new language, intended as a successor to the ABC language, would primarily replace the Unix shell and the C language for system administration. Guido himself was a big fan of the BBC television series * Monty Python's Flying Circus *, so he chose the word Python as the name of his new language.!
2. February 1991: Guido van Rossum posted the initial code for the Python interpreter on the alt.sources newsgroup, marked as version 0.9.0.
3. January 1994: Python 1.0 was released, the place where the dream began.
4. October 2000: Python 2.0 was released, making the entire Python development process more transparent and the ecosystem began to slowly take shape.
5. December 2008: Python 3.0 was released, introducing many new features of modern programming languages, but it was not fully backward compatible.
6. April 2011: pip was released for the first time, giving the Python language its own package management tool.
7. July 2018: Guido van Rossum announced a "permanent leave" from his position as "the benevolent dictator for life" (the person who has the final say in disputes within the open-source project community).
8. January 2020: After 11 years of coexistence between Python 2 and Python 3, official updates and maintenance of Python 2 have ceased, and users are encouraged to switch to Python 3 as soon as possible.
9. Currently, Python is widely used in large models (GPT-3, GPT-4, BERT, etc.), computer vision (image recognition, object detection, image generation, etc.), intelligent recommendation (YouTube, Netflix, ByteDance, etc.), autonomous driving (Waymo, Apollo, etc.), speech recognition, data science, quantitative trading, automated testing, automated operation and maintenance, and the Python language ecosystem is also quite prosperous.

# Python's advantages and disadvantages

Python has many advantages; here are a few examples.

1. Simple and elegant , Python is easier to learn than many other programming languages .
2. It allows you to do more with less code, improving development efficiency .
3. It is open source and has a strong community and ecosystem .
4. It can do a lot of things and has a very strong adaptability.
5. A glue language , capable of bonding together things developed in other languages.
6. Interpreted languages ​​are easier to cross-platform and can run on multiple operating systems.

Python's main drawback is its low execution efficiency (a common problem with interpreted languages). If you value code execution efficiency more, C, C++, or Go might be a better choice.

## Installing the Python Interpreter

Below, we'll use Windows 11 as an example to explain how to install a Python environment on the Windows operating system. Double-click to run the installer downloaded from the official website; an installation wizard will open, as shown in the image below.



![4.png](images/4.png)


### 1. Configure System Environment
On the initial installation screen, look for the checkbox at the bottom labeled **"Add python.exe to PATH"** and ensure it is checked.
* **Purpose:** This adds the Python interpreter to your Windows system's PATH environment variable.
* **Benefit:** It allows you to run `python` or `pip` commands from any terminal or command prompt without having to navigate to the installation folder.

### 2. Elevate Privileges
Select the option **"Use admin privileges when installing py.exe."**
* **Purpose:** This grants the installer administrator rights during the setup process.
* **Benefit:** It is highly recommended to check this to avoid permission issues when installing global packages or updating libraries later.

### 3. Choose Custom Installation
Instead of using the default "Install Now" option, select **"Customize Installation."**
* **The "Professional" Choice:** Using the default installation often places Python in deep, hidden folders (like `AppData`).
* **Benefit:** Custom installation allows you to choose a clean, easy-to-find directory (e.g., `C:\Python312`) and verify exactly which features are being installed.
