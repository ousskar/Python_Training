# While loops

The **while** loop repeats a block of code as long as a specified condition is true. It is useful when you do not know in advance how many times the loop should run.

````python
counter = 0

while counter < 5:
    print(counter)
    counter += 1  # Move toward the stopping condition.
````

## Simulating a do-while loop

Python does not have a `do...while` loop. When you need code to run at least once, use `while True` and exit with `break` when the stopping condition is met.

````python
attempt = 1

while True:
    print(f"Attempt {attempt}")
    if attempt >= 3:
        break  # Stop the infinite loop when the condition is met.
    attempt += 1
````

````python
min_length = 2
name = "A1"

while not (len(name) >= min_length and name.isprintable() and name.isalpha()):
    print(f"Rejected: {name}")
    name = "Alice"  # Simulate the next value a user might enter.

print(f"Hello, {name}")
````

````python
min_length = 2
names_to_try = ["", "Bo"]
index = 0

while True:
    name = names_to_try[index]
    index += 1

    if len(name) >= min_length and name.isprintable() and name.isalpha():
        break  # Exit as soon as a valid name is found.

print(f"Hello, {name}")
````

# Changing loop behavior with `break`, `continue`, and `pass`

- `break` stops the loop immediately.
- `continue` skips the rest of the current iteration and moves to the next one.
- `pass` does nothing. It is useful as a temporary placeholder where Python requires a block.

````python
number = 0

while number < 10:
    number += 1

    if number % 2 == 0:
        continue  # Skip even numbers.

    print(number)
````

## `while` loops with `else`

A loop `else` block runs only when the loop finishes normally. If the loop stops with `break`, the `else` block is skipped.

First, here is the manual approach: use a flag variable to remember whether the value was found.

````python
numbers = [1, 2, 3]
value = 4

found = False  # Track whether the search succeeds.
index = 0

while index < len(numbers):
    if numbers[index] == value:
        found = True
        break
    index += 1

if not found:
    numbers.append(value)

print(numbers)
````

The same search can be written with `while...else`. If the loop reaches the end without `break`, the `else` block appends the missing value.

````python
numbers = [1, 2, 3]
value = 4
index = 0

while index < len(numbers):
    if numbers[index] == value:
        break
    index += 1
else:
    numbers.append(value)  # Runs only if the loop did not break.

print(numbers)
````

# For loops

A `for` loop iterates over an iterable such as a list, tuple, string, dictionary, range, or generator. Use a `for` loop when you already have a collection of values to process.

```python
for item in iterable:
    # Do something with item
```

````python
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)
````

````python
for index in range(5):
    print(index)
````

````python
for character in "hello":
    print(character)
````

````python
colors = ["red", "green", "blue"]

for index, color in enumerate(colors, start=1):
    print(index, color)  # enumerate() gives both position and value.
````

````python
for number in range(3):
    print(number)
else:
    print("Finished without break")  # Runs because no break occurred.
````

````python
for number in range(5):
    if number == 3:
        break  # Stop before printing 3.
    print(number)
````

# Practical loop patterns

The next examples show common patterns you will use often: initializing variables, looping over related sequences, writing compact list transformations, and using `pass` as a placeholder.

## Initializing variables before a loop

A loop often updates a variable each time it runs. Create that variable before the loop starts so the starting value is explicit.

````python
total = 0  # Accumulator starts at zero.
numbers = [4, 7, 2]

for number in numbers:
    total += number

print(f"Total: {total}")
````

## Looping over two sequences with `zip()`

Use `zip()` when related values are stored in separate sequences and you want to process them together.

````python
students = ["Alice", "Bob", "Charlie"]
scores = [18, 15, 20]

for student, score in zip(students, scores):
    print(f"{student}: {score}")  # zip() pairs matching items.
````

## Single-line loops with list comprehensions

A list comprehension is a compact way to build a new list from an iterable. Use it when the transformation is simple and easy to read.

````python
numbers = [1, 2, 3, 4, 5]

squares = [number ** 2 for number in numbers]
even_numbers = [number for number in numbers if number % 2 == 0]  # Keep only even values.

print(squares)
print(even_numbers)
````

## Using `pass` as a placeholder

`pass` does nothing. It is useful when Python requires a block of code, but you have not written the real logic yet.

````python
for number in range(1, 6):
    if number < 3:
        pass  # Placeholder for logic we may add later.
    else:
        print(f"Ready to process: {number}")
````
