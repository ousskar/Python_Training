# 1. Inheritance

```python
class Person:
    def __init__(self, name):
        self.name = name

    def introduce(self):
        return f"Hello, I am {self.name}."

class Student(Person):
    pass
```

## 1.1 Overriding Methods and Using `super()`

A child class inherits behavior from its parent class, but it can also add new behavior or override existing methods. When you still want to reuse logic from the parent class, call it with `super()`.

```python
class Person:
    def __init__(self, name, person_id):
        self.name = name
        self.person_id = person_id

    def display_info(self):
        return f"Name: {self.name}, ID: {self.person_id}"
```

```python
class Student(Person):
    def __init__(self, name, person_id, program):
        super().__init__(name, person_id)
        self.program = program

    def display_info(self):
        base = super().display_info()
        return f"{base}, Program: {self.program}"

    def study(self):
        return f"{self.name} is studying {self.program}."
```

```python
student = Student("Samuel", 102, "Computer Science")
print(student.display_info())
print(student.study())
```

## 1.2 A Larger Inheritance Example

This example keeps the same parent-child relationship but adds extra state and overrides a method to provide richer output.

```python
class Person:
    def __init__(self, name, person_id):
        self.name = name
        self.person_id = person_id

    def get_details(self):
        return f"Name: {self.name}, ID: {self.person_id}"

class Student(Person):
    def __init__(self, name, person_id, grade, courses):
        super().__init__(name, person_id)
        self.grade = grade
        self.courses = courses

    def get_details(self):
        courses_text = ', '.join(self.courses)
        return f"Name: {self.name}, ID: {self.person_id}, Grade: {self.grade}, Courses: {courses_text}"

student = Student("Samuel", 5678, "B+", ["Math", "Physics", "Computer Science"])
print(student.get_details())
```

## 1.3 Multiple Inheritance and MRO

Python allows a class to inherit from more than one parent. When that happens, Python uses the method resolution order, or MRO, to decide which class to search first when looking up attributes and methods.

```python
class Person:
    def get_details(self):
        return "Details of a person."

class Athlete:
    def get_skill(self):
        return "Athletic skills."

class Student(Person, Athlete):
    pass

student = Student()
print(student.get_details())
print(student.get_skill())
```

```python
print(Student.__mro__)
```

# 2. Abstraction with Abstract Base Classes

Abstract base classes, often called ABCs, define a common interface that subclasses must implement. They are useful when you want several classes to share the same required methods but provide different implementations.

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass
```

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):  # Abstract Base Class
    def __init__(self, make, model):
        self.make = make
        self.model = model

    @abstractmethod
    def start_engine(self):
        pass

    @abstractmethod
    def stop_engine(self):
        pass

    def display_info(self):
        print(f"Make: {self.make}, Model: {self.model}")

class Car(Vehicle):
    def start_engine(self):
        print("Car engine started.")

    def stop_engine(self):
        print("Car engine stopped.")

class Motorcycle(Vehicle):
    def start_engine(self):
        print("Motorcycle engine started.")

    def stop_engine(self):
        print("Motorcycle engine stopped.")

# Usage
my_car = Car("Toyota", "Camry")
my_car.display_info()
my_car.start_engine()
my_car.stop_engine()

my_motorcycle = Motorcycle("Harley-Davidson", "Iron 883")
my_motorcycle.display_info()
my_motorcycle.start_engine()
my_motorcycle.stop_engine()
```
