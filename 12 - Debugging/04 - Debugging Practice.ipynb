## 1. Practice Scenario

The function below is supposed to print each author's average score. It contains a few assumptions that are safe for some inputs and risky for others.

```python
def calculate_average(scores):
    """Calculate the average of a list of scores."""
    if not scores:
        return 0
    return sum(scores) / len(scores)

def process_authors_scores(authors_scores):
    """Print each author's average score."""
    for author, scores in authors_scores.items():
        average = calculate_average(scores)
        print(f"Author: {author}, Average Score: {average:.2f}")

authors_scores = {
    "Alice": [85, 90, 78],
    "Bob": [92, 88],
    "Charlie": [70, 75, 80, 85],
    "Dana": []
}

process_authors_scores(authors_scores)
```

## 2. Diagnose the Issue

Questions to ask before changing the code:

- What happens when an author has no scores?
- Does `calculate_average([])` return a valid value?
- Which function should own the decision to skip or print empty results?

## 3. One Possible Fix

A common repair is to skip authors whose score lists are empty, or print a more explicit message for them.

```python
def calculate_average(scores):
    if not scores:
        return None
    return sum(scores) / len(scores)

def process_authors_scores(authors_scores):
    for author, scores in authors_scores.items():
        average = calculate_average(scores)
        if average is None:
            print(f"Author: {author}, no scores available")
            continue
        print(f"Author: {author}, Average Score: {average:.2f}")

process_authors_scores(authors_scores)
```

## 4. Second Exercise

This function should return the duplicate items in a list. It contains a bug in the line that records duplicates.

```python
def find_duplicates(items):
    duplicates = []
    seen = set()

    for item in items:
        if item in seen:
            duplicates.append[item]  # Intentional bug
        else:
            seen.add(item)

    return duplicates

find_duplicates([1, 2, 3, 2, 4, 3, 5])
```

## 5. Fixed Version

The bug is that `append` is a method call, not an index operation. Replacing square brackets with parentheses fixes the issue.

```python
def find_duplicates(items):
    duplicates = []
    seen = set()

    for item in items:
        if item in seen:
            duplicates.append(item)
        else:
            seen.add(item)

    return duplicates

find_duplicates([1, 2, 3, 2, 4, 3, 5])
```

## 6. Challenge Exercise

This final exercise is intentionally less obvious. The function below is supposed to summarize sales records by salesperson, but the input data contains several edge cases.

Your goal is to debug it step by step and make it robust enough to handle the provided dataset.

```python
def summarize_sales(records):
    totals = {}

    for record in records:
        salesperson = record["salesperson"].strip().title()
        amount = float(record["amount"])

        if salesperson not in totals:
            totals[salesperson] = []

        totals[salesperson] += amount

    for salesperson, amounts in totals.items():
        average = sum(amounts) / len(amounts)
        print(f"{salesperson}: count={len(amounts)}, total={sum(amounts):.2f}, average={average:.2f}")

sales_data = [
    {"salesperson": "alice", "amount": "120.5"},
    {"salesperson": " Bob ", "amount": "85"},
    {"salesperson": "alice", "amount": None},
    {"salesperson": "charlie", "amount": "not_available"},
    {"salesperson": "", "amount": "40"},
    {"salesperson": "bob", "amount": "95.5"},
    {"amount": "70"},
    {"salesperson": "alice", "amount": "110"},
]

summarize_sales(sales_data)
```

### Debugging Goals

Try to identify and fix at least these categories of problems:

- missing keys or empty salesperson names
- invalid numeric values such as `None` or non-numeric text
- incorrect accumulation logic when storing totals
- assumptions that break the summary loop at the end
