2. Common Error Types

**SyntaxError**

Occurs when Python cannot interpret your code due to incorrect syntax.

```python
# Example
print("Hello World'  # Missing closing quote
```

**NameError**

Occurs when a variable is not defined.

```python
# Example
print(undefined_variable)
```

**TypeError**

Occurs when an operation is performed on an inappropriate type.

```python
1 + "a"
```

```python
# Example
my_list = [1, 2, 3]
print(my_list[3])  # Raises IndexError because valid indexes are 0, 1, and 2
```

**KeyError**

Occurs when a dictionary key doesn't exist.

```python
# Example
my_dict = {'a': 1, 'b': 2}
print(my_dict['c'])
```

3. Using print() for Basic Debugging

```python
def calculate_average(numbers):
    print(f"Input numbers: {numbers}")  # Debug print
    total = sum(numbers)
    print(f"Total: {total}")  # Debug print
    average = total / len(numbers)
    print(f"Average: {average}")  # Debug print
    return average

calculate_average([1, 2, 3, 4])
```

4. Using Built-in Functions for Debugging

**type()**

Check the type of a variable.

```python
x = "5"
print(type(x))
```

**id()**

Get the memory address of an object.

```python
a = [1, 2, 3]
b = a
print(id(a), id(b))  # Same id means they reference the same object
```

**dir()**

List the available attributes and methods of an object.

```python
sample_text = "debugging"
print(dir(sample_text)[:10])  # Show a few available string attributes and methods
```

**vars()**

Return the dict attribute of an object.

```python
class MyClass:
    def __init__(self):
        self.x = 1
        self.y = 2

obj = MyClass()
print(vars(obj))
```

**help()**

Access Python's built-in documentation.

```python
help(str.split)
```
