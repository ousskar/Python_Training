# 2. Functions and Files Exercises Solutions

This notebook contains sample solutions for the functions, scope, type hints, file handling, CSV, JSON, and module exercises.

```python
def greet(name):
    return f'Hello, {name}!'

def square(number):
    return number * number

def is_even(number):
    return number % 2 == 0

def area_rectangle(width, height):
    return width * height

def full_name(first, last):
    return f'{first} {last}'

def describe_person(name, country='Unknown'):
    return f'{name} is from {country}'

def sum_all(*values):
    return sum(values)

def print_kwargs(**kwargs):
    for key, value in kwargs.items():
        print(key, value)
```

```python
global_message = 'hello'

def read_global():
    return global_message

def shadow_global():
    global_message = 'local hello'
    return global_message

def multiply(a: int, b: int) -> int:
    return a * b

def average(values: list[float]) -> float:
    return sum(values) / len(values)

def maybe_average(values: list[float]) -> float | None:
    if not values:
        return None
    return sum(values) / len(values)
```
