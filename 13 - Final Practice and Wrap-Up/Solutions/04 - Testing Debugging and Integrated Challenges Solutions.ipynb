```python
import unittest
from unittest.mock import Mock

def add_vat(price, rate=0.2):
    if price < 0:
        raise ValueError('price must be non-negative')
    return price * (1 + rate)

class TestAddVat(unittest.TestCase):
    def setUp(self):
        self.base_price = 100

    def test_add_vat(self):
        self.assertEqual(add_vat(100), 120)

    def test_invalid_price(self):
        with self.assertRaises(ValueError):
            add_vat(-5)

class TestMultipleValues(unittest.TestCase):
    def test_even_values(self):
        for value in [2, 4, 6]:
            with self.subTest(value=value):
                self.assertEqual(value % 2, 0)

mock_sender = Mock()
mock_sender.send('a@example.com', 'hello')
mock_sender.send.assert_called_once_with('a@example.com', 'hello')
```

```python
from pathlib import Path
import json
from functools import wraps

def read_names_title_case(file_path):
    lines = Path(file_path).read_text(encoding='utf-8').splitlines()
    return [line.title() for line in lines if line.strip()]

class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def to_dict(self):
        return {'name': self.name, 'price': self.price}

def log_calls(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        print(f'Calling {func.__name__}')
        return func(*args, **kwargs)
    return wrapper

@log_calls
def average_price(products):
    prices = [product['price'] for product in products]
    return sum(prices) / len(prices)

products = [Product('Pen', 2.5).to_dict(), Product('Notebook', 5.0).to_dict()]
json_text = json.dumps(products)
loaded_products = json.loads(json_text)
print(average_price(loaded_products))
```
