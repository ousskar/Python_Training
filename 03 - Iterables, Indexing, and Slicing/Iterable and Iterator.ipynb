# 1. Iterables and Iterators

## 1. What Is an Iterable?

An iterable is any Python object that can be looped over one item at a time. Common examples include strings, lists, tuples, dictionaries, and sets.

You can pass an iterable to `iter()` to create an iterator.

```python
my_list = [1, 2, 3]
for item in my_list:
    print(item)  # Under the hood, Python uses iter() and next()
```

## 2. How to Check If an Object Is Iterable

```python
from collections.abc import Iterable

is_iterable = isinstance([1, 2, 3], Iterable)  # True
print(is_iterable)
```

## 3. What Is an Iterator?

An iterator is an object that keeps track of where it is during iteration.

It implements:

- `__iter__()` to return the iterator object itself
- `__next__()` to return the next value
- `StopIteration` to signal that there are no more values

```python
my_list = [10, 20, 30]
my_iter = iter(my_list)

print(next(my_iter))  # 10
print(next(my_iter))  # 20
print(next(my_iter))  # 30

# Calling next(my_iter) again would raise StopIteration
```

```python
my_iter = iter([100, 200, 300])
print(next(my_iter))
```

```python
my_iter = iter([1, 2, 3])

while True:
    try:
        item = next(my_iter)
        print(item)
    except StopIteration:
        break
```

## 4. Creating Your Own Iterator Class

```python
class Countdown:
    def __init__(self, start):
        self.current = start

    def __iter__(self):
        return self

    def __next__(self):
        if self.current <= 0:
            raise StopIteration
        value = self.current
        self.current -= 1
        return value

for number in Countdown(5):
    print(number)
```

## 5. When to Use an Iterator in Python

✅ 1. You Want to Process Items One at a Time

If you're dealing with large datasets (e.g., reading from a file or streaming data), iterators let you process elements lazily — one at a time — without loading everything into memory.

```python
# A file object is iterable, but this notebook uses a list of lines
# so the example stays runnable without external files.
lines = ["first line\n", "second line\n", "third line\n"]

for line in lines:
    print(line.strip())
```

✅ 2. You Need to Implement Custom Iteration Logic

When your iteration logic is not a simple loop (e.g., skip values, dynamic stop conditions), defining a custom iterator class gives you full control.

```python
class EvenNumbers:
    def __init__(self, max_num):
        self.num = 0
        self.max = max_num

    def __iter__(self):
        return self

    def __next__(self):
        if self.num > self.max:
            raise StopIteration
        current = self.num
        self.num += 2
        return current

for n in EvenNumbers(10):
    print(n)
```

✅ 3. You Need Lazy Evaluation (Efficiency)

Use iterators to compute values only when needed, which improves performance and saves memory.

```python
def count_up():
    num = 0
    while True:
        yield num

        num += 1

gen = count_up()
print(next(gen))  # 0
print(next(gen))  # 1
```

✅ 4. You're Working with Infinite Sequences

Iterators (especially generators) are ideal for creating sequences that can go on forever (e.g., Fibonacci, counters, streams).

✅ 5. You Want to Abstract Complex Sequences

Iterators let you define elegant, reusable components that encapsulate complex state and logic.

## 6. Summary

| Term        | Meaning                                      |
| ----------- | -------------------------------------------- |
| Iterable    | An object that can return an iterator        |
| Iterator    | An object with `__iter__()` and `__next__()` |
| `iter(obj)` | Gets an iterator from an iterable            |
| `next(obj)` | Gets the next item from an iterator          |


# 2. Indexing and Slicing

Iterables like strings, lists, and tuples are often also **sequences**, which means you can access elements by position and extract parts of them with slicing.

## 2.1 Accessing Elements with Indexing

Indexing starts at `0`. Negative indexes count from the end of the sequence.

```python
text = "Python"
numbers = [10, 20, 30, 40]
coordinates = (5, 8)

print(text[0])
print(text[-1])
print(numbers[1])
print(coordinates[0])
```

## 2.2 Extracting Parts with Slicing

A slice uses the form `sequence[start:stop]`. You can also provide a third value, `step`, to skip items or reverse a sequence.

```python
text = "Iteration"
numbers = [0, 1, 2, 3, 4, 5]

print(text[1:5])
print(text[:4])
print(text[::2])
print(text[::-1])

print(numbers[2:5])
print(numbers[:3])
print(numbers[::-1])
```

## 2.3 A Quick Connection to Loops

Indexing and slicing are useful on their own, but they also help you understand what loops are iterating over and which part of a sequence you want to process.

```python
colors = ["red", "green", "blue"]

for index, color in enumerate(colors):
    print(index, color)
```

## 2.4 Bridge to Loop Helpers

Once you understand iterables and indexing, helpers like `enumerate()` become much easier to read: they give both the position and the value while looping.

## 2.5 Advanced Iterable Tricks

These examples go one step further than the basics. They introduce patterns that are common in real Python code and prepare you for more expressive loops later.

### 1. `enumerate()` with a custom start

You can ask `enumerate()` to begin counting from a value other than `0`.

```python
items = ["apple", "banana", "cherry"]

for index, item in enumerate(items, start=1):
    print(index, item)
```

### 2. Iterating over multiple sequences with `zip()`

`zip()` groups values from multiple iterables together so they can be processed side by side.

```python
names = ["Alice", "Bob", "Charlie"]
scores = [15, 18, 12]

for name, score in zip(names, scores):
    print(name, score)
```

### 3. Iterator exhaustion

An iterator is a one-pass object. Once you consume its values, there is nothing left to read unless you create a new iterator.

```python
numbers = iter([1, 2, 3])

for number in numbers:
    print(number)

print(list(numbers))
```

### 4. Generator expressions

A generator expression creates values lazily, one at a time, instead of building the whole result immediately.

```python
squares = (x * x for x in range(5))

print(next(squares))
print(next(squares))
print(list(squares))
```

### More tools for advanced iteration

The next examples continue the same advanced section with tools from `itertools`, sequence unpacking, and compact loop-style checks.

### 5. `itertools.chain` and `itertools.islice`

`chain()` joins multiple iterables into one sequence. `islice()` takes a slice of an iterable without needing to convert it to a list.

```python
from itertools import chain, islice

a = [1, 2, 3]
b = [4, 5, 6]

print(list(chain(a, b)))
print(list(islice(range(100), 5, 10)))
```

### 6. Transposing rows and columns with `zip(*matrix)`

Combining `zip()` with argument unpacking is a compact way to swap rows and columns.

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
]

transposed = list(zip(*matrix))
print(transposed)
```

### 7. Quick checks with `any()` and `all()`

Combined with a generator expression, `any()` and `all()` express a full loop check in one line.

```python
numbers = [2, 4, 6, 7, 10]

print(any(n > 8 for n in numbers))
print(all(n % 2 == 0 for n in numbers))
```

### 8. Unpacking with `*rest`

Extended unpacking lets you capture the first and last items separately from everything else.

```python
values = [10, 20, 30, 40, 50]

first, *middle, last = values
print(first)
print(middle)
print(last)
```
