# Variables in Python

## What is a variable?

A **variable** is a name bound to a value. In Python, variables do not need an explicit type declaration. You assign a value with `=` and Python keeps track of the value's type for you.

## Assigning values to variables

When you write `name = "Alice"`, the variable `name` now refers to the string value `"Alice"`. You can later reassign the same variable to a different value.

```python
# Variable assignments
name = "Alice"
age = 25
height = 1.68
is_student = True


# Display the variables
print(f"Name: {name}, Type: {type(name)}")
print(f"Age: {age}, Type: {type(age)}")
print(f"Height: {height}, Type: {type(height)}")
print(f"Is Student: {is_student}, Type: {type(is_student)}")
```

## Reassignment and dynamic typing

A variable can be reassigned to a new value later. In Python, the **name** stays the same, but the **value and type** can change.

The next cell shows reassignment in action. `student_count` starts as an integer, then the same variable name is reused for a string to demonstrate Python's dynamic typing.

```python
student_count = 25
print(student_count)
print(type(student_count))

student_count = "twenty-five"
print(student_count)
print(type(student_count))
```

## Variables in expressions

Variables become useful when they participate in calculations. In the next cell, `length` and `width` are reused to compute several related values from the same inputs.

## Multiple assignment and unpacking

Python lets you assign several variables in one line. This is common when unpacking values from a tuple or list.

```python
# Define variables
length = 10
width = 5

area = length * width
perimeter = 2 * (length + width)
diagonal = (length**2 + width**2)**0.5

print(f"Area: {area}")
print(f"Perimeter: {perimeter}")
print(f"Diagonal: {diagonal:.2f}")
```

```python
x, y = 10, 20
first_name, last_name = "Ada", "Lovelace"

print(x, y)
print(first_name, last_name)
```

# Variable Naming Rules

- Must start with a **letter** or an **underscore** (`_`)
- Can contain **letters**, **numbers**, and **underscores** (`_`)
- Cannot contain spaces or symbols like `-`, `@`, or `!`
- Are **case-sensitive**: `age`, `Age`, and `AGE` are different names
- Cannot use **Python keywords** such as `if`, `for`, `class`, or `def`
- Avoid shadowing built-ins like `list`, `str`, or `dict` even though Python allows it

## Checking reserved keywords

Python keywords are reserved words that already have a meaning in the language. You cannot use them as variable names, so it is useful to know how to inspect the built-in keyword list.

```python
# Python keywords
import keyword
print(keyword.kwlist)
```

## Constants and placeholder values

Python does not have true constants, but by convention we write constant names in uppercase. `None` is commonly used when a variable has no value yet.

```python
valid_name = "Alice"
_private_name = "internal"
student2 = "Bob"

print(valid_name)
print(_private_name)
print(student2)
```

```python
MAX_RETRIES = 3
selected_user = None

print(MAX_RETRIES)
print(selected_user)
print(type(selected_user))
```

Use uppercase names for constants by convention, for example `MAX_RETRIES = 3`.

## Useful variable tricks

These small patterns are common in real Python code. They are worth learning early because they make code shorter, clearer, and easier to debug.

### 1. Swapping values without a temporary variable

Python lets you swap two variables in one line using tuple unpacking.

```python
a = 10
b = 20
a, b = b, a

print(a, b)
```

### 2. Augmented assignment

Operators like `+=`, `-=`, and `*=` update a variable using its current value.

```python
count = 0
count += 1
count += 4

print(count)
```

### 3. Checking for `None` the Pythonic way

Use `is None` and `is not None` when checking whether a variable has no value yet.

```python
user = None

if user is None:
    print("No user selected yet")
```

### 4. Debug-style f-strings

When you are learning or debugging, `f"{variable=}"` prints both the variable name and its value.

```python
total = 42
name = "Alice"

print(f"{total=}")
print(f"{name=}")
```

### 5. One name or two names? Shared references

For mutable values like lists, two variables can point to the same object. Changing one name can affect the other.

To avoid that, create a copy instead of assigning the same list to a second variable.

- `alias = numbers` means both names point to the same list
- `alias = numbers.copy()` creates a new list with the same values

For nested data structures, a shallow copy may not be enough. In that case, use `copy.deepcopy()`.

```python
# Shared reference: both names point to the same list
numbers = [1, 2, 3]
alias = numbers
alias.append(4)

print("shared reference:", numbers, alias)

# Safe copy: each name refers to a different list
numbers = [1, 2, 3]
alias = numbers.copy()
alias.append(4)

print("copied list:", numbers, alias)
```
