# 1. Custom Classes

In this notebook, you will build simple custom classes and gradually add more Python behavior to them. The goal is to understand how objects store data, expose methods, and interact with built-in operations such as `str()`, comparison operators, and properties.

## 1.1 Defining a Class

To create a custom class, use the `class` keyword. The special method `__init__` initializes the attributes of each new instance.

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
```

We create **instances** of `Rectangle` by calling the class with the values needed by `__init__`. Python automatically passes the new object as the first argument, traditionally named `self`.

Using `self` is a convention, but it is also the standard style you should follow. Technically you can choose another valid name, but doing so usually makes the code harder to read.

```python
r1 = Rectangle(10, 20)
r2 = Rectangle(3, 5)
```

```python
r1.width
```

```python
r2.height
```

## 1.2 Attributes and Methods

`width` and `height` are attributes of each `Rectangle` instance. Since they store values rather than functions, they are data attributes. Callable attributes defined inside the class are methods.

You access attributes and methods with dot notation, such as `r1.width` or `r1.area()`. The expression before the dot identifies the object, and the name after the dot identifies the attribute or method you want to use.

## 1.3 Adding Instance Methods

We can add callable attributes to the class as methods. Instance methods receive the object itself as the first parameter, which lets them read or update that object's state.

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(the_referenced_object):
        return 2 * (the_referenced_object.width + the_referenced_object.height)
```

```python
r1 = Rectangle(10, 20)
```

```python
r1.area()
```

When you call `r1.area()`, Python automatically passes `r1` as the first argument to the method. That is why instance methods can access the state stored on the object through `self`.

This also explains why the first parameter does not have to be named `self`. Python passes the object positionally. Even so, using `self` is the standard and should be preferred in real code.

```python
r1.perimeter()
```

The example works, but it is intentionally non-idiomatic. Treat it as a demonstration of how method binding works, not as a naming style to copy.

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)
```

```python
r1 = Rectangle(10, 20)
```

# 2. Special Methods

Python provides many **special methods** that let custom classes behave more like built-in objects. These methods control behavior such as string conversion, comparison, and operator overloading.

For example, we can obtain the string representation of an integer using the built-in `str` function:

```python
str(10)
```

What happens if we try this with our Rectangle object?

```python
str(r1)
```

The default result is not very informative. Python has no way to guess how your rectangle should be displayed, so you must define that behavior yourself if you want a clearer string form.

One option would be to write a regular method such as `to_str()`, but special methods give you a more natural integration with built-in functions.

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

    def to_str(self):
        return 'Rectangle (width={0}, height={1})'.format(self.width, self.height)
```

So now we could get a string from our object as follows:

```python
r1 = Rectangle(10, 20)
r1.to_str()
```

But of course, using the built-in `str` function still does not work:

```python
str(r1)
```

That approach works, but it is not the most Pythonic design. If every class invented a different method name for string conversion, users of the class would have to memorize custom conventions instead of relying on built-in behavior.

## 2.1 `__str__` and `__repr__`

When you call `str(r1)`, Python checks whether the class defines `__str__`. If it does, that method returns the user-friendly string form of the object.

Related to this is `__repr__`, which is the representation Python often uses in interactive environments such as Jupyter.

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

    def __str__(self):
        return 'Rectangle (width={0}, height={1})'.format(self.width, self.height)
```

```python
r1 = Rectangle(10, 20)
```

```python
str(r1)
```

However, in Jupyter (and interactive console if you are using that), look what happens here:

```python
r1
```

In this context Python is not calling `str(r1)`. Jupyter is asking for the object's representation, so Python looks for `__repr__`. That is why defining only `__str__` improves `print(r1)` but does not change how the bare object is displayed in a notebook cell.

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

    def __str__(self):
        return 'Rectangle (width={0}, height={1})'.format(self.width, self.height)

    def __repr__(self):
        return 'Rectangle({0}, {1})'.format(self.width, self.height)
```

```python
r1 = Rectangle(10, 20)
```

```python
print(r1)  # uses __str__
```

```python
r1  # uses __repr__
```

## 2.2 Equality with `__eq__`

Comparison with `==` does not automatically compare the internal values of two custom objects. If you want two rectangles with the same dimensions to count as equal, you need to define that rule explicitly with `__eq__`.

```python
r1 = Rectangle(10, 20)
r2 = Rectangle(10, 20)
```

```python
r1 == r2
```

By default, `==` compares object identity unless the class defines its own equality rule. Python does not automatically know that two rectangles with the same dimensions should be treated as equal values.

We just need to tell Python how to do it, using the special method `__eq__`.

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

    def __str__(self):
        return 'Rectangle (width={0}, height={1})'.format(self.width, self.height)

    def __repr__(self):
        return 'Rectangle({0}, {1})'.format(self.width, self.height)

    def __eq__(self, other):
        print('self={0}, other={1}'.format(self, other))
        if isinstance(other, Rectangle):
            return (self.width, self.height) == (other.width, other.height)
        return NotImplemented
```

```python
r1 = Rectangle(10, 20)
r2 = Rectangle(10, 20)
```

```python
r1 is r2
```

```python
r1 == r2
```

```python
r3 = Rectangle(2, 3)
```

```python
r1 == r3
```

And if we try to compare our Rectangle to a different type:

```python
r1 == 100
```

The `print` call inside `__eq__` is only there to make the method call visible while learning. In normal code, comparison methods should avoid side effects and simply return a result.

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

    def __str__(self):
        return 'Rectangle (width={0}, height={1})'.format(self.width, self.height)

    def __repr__(self):
        return 'Rectangle({0}, {1})'.format(self.width, self.height)

    def __eq__(self, other):
        if isinstance(other, Rectangle):
            return (self.width, self.height) == (other.width, other.height)
        else:
            return False
```

## 2.3 Ordering with `__lt__`

Python can also compare custom objects with operators such as `<` and `>`, but only if you define the corresponding special methods. In this example, we compare rectangles by area.

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

    def __str__(self):
        return 'Rectangle (width={0}, height={1})'.format(self.width, self.height)

    def __repr__(self):
        return 'Rectangle({0}, {1})'.format(self.width, self.height)

    def __eq__(self, other):
        if isinstance(other, Rectangle):
            return (self.width, self.height) == (other.width, other.height)
        else:
            return False

    def __lt__(self, other):
        if isinstance(other, Rectangle):
            return self.area() < other.area()
        else:
            return NotImplemented
```

```python
r1 = Rectangle(100, 200)
r2 = Rectangle(10, 20)
```

```python
r1 < r2
```

```python
r2 < r1
```

What about `>`?

```python
r1 > r2
```

Why did `r1 > r2` work even though we did not define `__gt__`? Python tried the reflected comparison `r2 < r1`. Since `__lt__` is defined, Python was able to complete the comparison using that method.

That fallback does not cover every comparison operator. For example, `<=` still needs its own implementation, such as `__le__`, unless you use a helper like `functools.total_ordering`.

```python
r1 <= r2
```

# 3. Managed Attributes with `@property`

If you come from a Java background, direct attribute access such as `width` and `height` may seem risky. In Python, it is common to start with direct attribute access and later introduce managed attributes with `@property` if validation or computed behavior becomes necessary.

Let's take our Rectangle class once again. I'll use a simplified version to keep the code short.

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def __repr__(self):
        return 'Rectangle({0}, {1})'.format(self.width, self.height)
```

```python
r1 = Rectangle(10, 20)
```

```python
r1.width
```

```python
r1.width = 100
```

```python
r1
```

So far, we can both read and assign `width` directly. That is convenient, but later you may want to enforce rules such as requiring positive values.

In some languages, changing from direct attribute access to getters and setters can force you to rewrite calling code. Python's `@property` pattern avoids that problem because users can keep writing `obj.width` even after you add validation.

## 3.1 Validation with Properties

In Python, `@property` lets you keep the clean `obj.width` syntax while still controlling how values are read and written. This gives you a way to add validation without breaking the code that already uses the class.

```python
class Rectangle:
    def __init__(self, width, height):
        self._width = width
        self._height = height

    def __repr__(self):
        return 'Rectangle({0}, {1})'.format(self.width, self.height)

    @property
    def width(self):
        return self._width

    @width.setter
    def width(self, width):
        if width <= 0:
            raise ValueError('Width must be positive.')
        self._width = width

    @property
    def height(self):
        return self._height

    @height.setter
    def height(self, height):
        if height <= 0:
            raise ValueError('Height must be positive.')
        self._height = height
```

```python
r1 = Rectangle(10, 20)
```

```python
r1.width
```

```python
r1.width = 100
```

```python
r1
```

```python
r1.width = -10
```

## 3.2 Validate During Initialization

Validation should happen not only when an attribute is changed later, but also when the object is first created. A good pattern is to route initialization through the same property setters so the rules stay in one place.

```python
class Rectangle:
    def __init__(self, width, height):
        self._width = None
        self._height = None
        # now we call our accessor methods to set the width and height
        self.width = width
        self.height = height

    def __repr__(self):
        return 'Rectangle({0}, {1})'.format(self.width, self.height)

    @property
    def width(self):
        return self._width

    @width.setter
    def width(self, width):
        if width <= 0:
            raise ValueError('Width must be positive.')
        self._width = width

    @property
    def height(self):
        return self._height

    @height.setter
    def height(self, height):
        if height <= 0:
            raise ValueError('Height must be positive.')
        self._height = height
```

```python
r1 = Rectangle(0, 10)
```

In a production version, you would usually add more checks, such as validating numeric types as well as positive values. The important idea is to fail early during object creation so errors appear at the line that creates the object, which makes debugging easier.

This notebook only introduces a small subset of Python's special methods. Later topics such as inheritance, dataclasses, and richer operator behavior will build on the same ideas.

## 4. Key Takeaways

- classes bundle data and behavior together
- instance methods receive the current object as their first argument
- special methods such as `__str__`, `__repr__`, `__eq__`, and `__lt__` let custom classes work naturally with Python syntax
- `@property` lets you add validation without changing how users access an attribute
- validating early during object creation usually makes bugs easier to find
